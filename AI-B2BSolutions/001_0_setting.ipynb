## 🚀 Step 3. 내 첫 번째 AI 코드 실행하고 결과 확인하기

준비가 끝났습니다! 이제 코랩 중앙에 있는 회색 코드 셀(Cell)에 아래 코드를 복사해서 붙여넣고 첫 AI 결과물을 뽑아봅시다.



`````python


# 1. 필요한 AI 도구(라이브러리) 설치 및 불러오기
!pip install -q openai


from google.colab import userdata
from openai import OpenAI

# 2. 안전하게 저장한 API Key 불러와서 AI 클라이언트 깨우기
# (만약 여기서 에러가 난다면 Step 2의 열쇠 토글스위치가 켜졌는지 확인하세요!)
api_key = userdata.get('OPENAI_API_KEY')
client = OpenAI(api_key=api_key)

# 3. AI에게 질문 던지기
response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "user", "content": "안녕하세요! 오늘부터 AI 강사 마스터 과정을 시작합니다. 응원 한마디 해주세요!"}
    ]
)

# 4. 결과 출력
print("🤖 AI의 답변:")
print(response.choices[0].message.content)
`````

## 📑 Section 1. OpenAI API Core & 파라미터 딥다이브

### 1. 구조적 프롬프팅을 위한 Message Role 분담

웹 브라우저로 접속하는 ChatGPT와 달리, 개발 환경(API)에서는 대화의 뼈대를 강사가 직접 구조화해야 합니다. API는 메시지를 다음 3가지 역할 구조로 기억합니다.

| Role (역할) | 핵심 정의 | 실무 활용 및 강사 팁 |
| --- | --- | --- |
| **`system`** (시스템 지시문) | • 모델에게 부여하는 **'최상위 헌법'**<br>• AI의 가치관, 말투, 전문 영역, 절대 하지 말아야 할 제약 조건을 강제 | 기업 교육 필수 패턴: <br>*"너는 우리 회사 규정만 답변하는 인사과 대리야. 외부 정보는 모른다고 답해."* 형태로 환각(Hallucination)을 방지할 때 씁니다. |
| **`user`** (사용자 입력) | • 실무자나 고객이 매번 새롭게 입력하는 **프롬프트 원본**<br>• 질문, 분석 요청 데이터 등이 이 역할을 통해 주입됨 | 현업에서 인입되는 이메일 전문이나 VOC(고객의 소리) 텍스트를 파이썬 변수로 묶어 이 자리에 매핑합니다. |
| **`assistant`** (AI 답변 기록) | • AI가 직전에 출력했던 **대화의 역사(History)**<br>• LLM은 과거를 기억 못 하므로, 이 기록을 누적해서 다시 보내야 맥락이 유지됨 | **"대화가 이어지게 만드는 마법의 키"**<br>3주차에 배울 멀티턴(Multi-turn) 챗봇 아키텍처의 핵심 뼈대가 됩니다. |

### 🎛️ 2. 현업에서 알아야 할 핵심 하이퍼파라미터

코드에 들어가는 숫자 하나가 비용과 품질, 그리고 시스템의 안정성을 결정합니다.

* **`temperature` (설정 범위: 0.0 ~ 2.0):**
* **개념:** 다음 단어를 선택할 때 확률 분포의 '예측 가능성(창의성)'을 조절합니다.
* **조절 팁:** 값이 **0.0에 가까울수록** 가장 확률이 높은 단어만 선택하므로 정확하고 일관된 답변(보고서 작성, 코드 생성, 데이터 분석)이 나오며, **1.0 이상으로 갈수록** 확률이 낮은 단어도 섞여 나와 창의적인 답변(마케팅 카피라이팅, 아이디어 브레인스토밍)이 나옵니다. *기업 실무 실습에서는 보통 0.0 ~ 0.3을 권장합니다.*


* **`max_tokens`:**
* **개념:** AI가 생성할 수 있는 최대 토큰(단어/글자 단위 조각) 수를 제한합니다.
* **조절 팁:** 무한정 답변이 길어져 비용이 폭발하거나 타임아웃 오류가 발생하는 것을 방지하는 방어벽 역할을 합니다.


* **`response_format` (최신 API 핵심 스펙):**
* **개념:** AI의 출력 형식을 완전히 강제하는 기능입니다. `{"type": "json_object"}`로 지정하면 AI는 다른 부가 설명(예: "네, 분석해 드리겠습니다.")을 완전히 생략하고 100% 순수한 JSON 데이터만 반환합니다.

### 💻 3. [실습 1] 페르소나 통제 및 데이터 요약/JSON 객체 파싱 자동화 (고도화)

> **시나리오:** 대량의 비즈니스 고객 불만 사항(VOC) 데이터를 받아, 인공지능이 임의로 포맷을 깨뜨리지 못하도록 **JSON 구조 출력을 완벽히 보장**하고, 이를 파이썬 객체(`dict`)로 변환하여 실무 정형 데이터 시스템에 즉시 연동 가능한 자동화 파이프라인을 구축합니다.


`````python

# 1. 필수 라이브러리 설치 (최초 1회 실행)
!pip install openai

import os
import json  # [수정 보완] AI가 출력한 JSON 문자열을 파이썬 객체로 변환하기 위한 라이브러리
from google.colab import userdata
from openai import OpenAI

# Google Colab 왼쪽 '열쇠' 모양 아이콘(Secrets)에 OPENAI_API_KEY가 등록되어 있어야 합니다.
client = OpenAI(api_key=userdata.get('OPENAI_API_KEY'))

# 분석할 원본 VOC 데이터 (실무 예시 예제)
raw_voc = """
어제 주문한 상품이 오늘 오전까지 온다고 해서 연차 쓰고 기다렸는데 오후 4시가 넘어서야 도착했습니다.
고객센터는 통화 연결도 안 되고, 배송 지연에 대한 아무런 안내 문자도 없었네요.
제품 자체는 만족스러운데 서비스 대응이 너무 최악이라 실망이 큽니다. 환불하고 싶을 정도예요.
"""

# API 호출 수행
response = client.chat.completions.create(
    model="gpt-4o-mini",      # 가성비가 가장 뛰어나 실습용으로 최적인 모델
    temperature=0.0,          # 데이터 정밀 분석을 위해 일관성 최대로 설정
    max_tokens=500,
    # [수정 보완] 마크다운 감싸기 없이 완벽한 JSON 출력을 보장하는 최신 API 스펙 필수 적용
    response_format={"type": "json_object"},
    messages=[
        {
            "role": "system",
            "content": (
                "너는 대기업 CS 센터의 데이터 분석 전문가야. 입력되는 고객 불만(VOC)을 분석해야 해.\n"
                "반드시 아래의 지시사항을 준수하여 JSON 포맷으로만 답변해라. 다른 부가 설명이나 텍스트는 절대 금지한다.\n\n"
                "JSON 구조 정의:\n"
                "- sentiment: '긍정', '중립', '부정' 중 하나\n"
                "- issue: 고객이 불만을 가진 핵심 요인을 3글자 내외의 키워드로 추출\n"
                "- summary: 현상을 요약한 한 줄 문장 (반드시 존댓말 형태)"
            )
        },
        {
            "role": "user",
            "content": f"다음 VOC를 분석해줘:\n{raw_voc}"
        }
    ]
)

# 1. AI가 출력한 순수 데이터 확인
json_string = response.choices[0].message.content
print("[1단계: AI가 반환한 순수 JSON 문자열]")
print(json_string)
print("-" * 50)

# 2. [수정 보완] 문자열 데이터를 파이썬 딕셔너리로 파싱하여 실무 정형화 연동 실습
parsed_data = json.loads(json_string)

print("[2단계: 파이썬 객체 변환 및 실무 데이터 추출 실습]")
print(f"✔ 감정 상태 분석 결과 : {parsed_data['sentiment']}")
print(f"✔ 불만 원인 핵심 요인 : {parsed_data['issue']}")
print(f"✔ 사내 공유용 요약본  : {parsed_data['summary']}")
`````



#### 💡 강사의 Line-by-Line 해설

* `response_format={"type": "json_object"}`: 이 파라미터가 누락되면 AI가 간혹 ````json`과 같은 불필요한 마크다운 문법을 섞어 출력하게 되며, 이는 자동화 시스템 구축 시 코드 에러를 유발하는 주범이 됩니다. 실무형 자동화를 교육할 때는 필수 조건입니다.
* `json.loads(json_string)`: AI가 준 데이터는 겉보기에만 JSON인 '글자(String)' 상태입니다. 파이썬이 이 데이터를 가지고 엑셀에 넣거나 이메일을 자동으로 쏘는 연동 작업을 수행하려면 반드시 `json.loads()`를 거쳐 파이썬 객체인 딕셔너리(`dict`)로 변환해 주어야 합니다.

### 💻 2. [실습 2] Azure OpenAI SDK 호출 실전

> **중요:** Azure 환경은 모델 이름 대신, Azure 포털에서 강사 또는 사내 관리자가 직접 지정한 **'배포 명칭(Deployment Name)'**을 매칭해야 정상 작동하며, 초반 셋업 가이드에 따라 `AZURE_OPENAI_KEY`가 구글 코랩에 연동되어 있어야 합니다.


`````python
from openai import AzureOpenAI
from google.colab import userdata

# 강사가 수업 전에 빌드하여 제공하는 고유 엔드포인트와 키 세팅
# (Step 2에서 등록해 둔 AZURE_OPENAI_KEY를 시스템이 자동으로 호출합니다)
client = AzureOpenAI(
    api_key = userdata.get('AZURE_OPENAI_KEY'),
    api_version = "2024-02-15-preview",
    azure_endpoint = "https://your-company-resource.openai.azure.com/" # 기업 리소스 주소
)

response = client.chat.completions.create(
    model="prod-gpt-4o-mini-deploy", # 포털에서 설정한 배포명 입력
    messages=[
        {"role": "system", "content": "너는 사내 보안 규정을 준수하는 안전한 AI 어시스턴트이다."},
        {"role": "user", "content": "사내 망에서 AI API를 사용할 때 주의해야 할 점 2가지를 알려줘."}
    ]
)

print(response.choices[0].message.content)
`````

## ⚠️ 강사 전용 트러블슈팅 매뉴얼 (수강생 빈출 에러 셋업)

1. **`AuthenticationError: No API key provided...`**
* **원인:** Colab의 Secrets 탭(열쇠 아이콘)에 키 이름(`OPENAI_API_KEY` 또는 `AZURE_OPENAI_KEY`)을 입력할 때 오타가 났거나, 우측의 '노트북 액세스 권한' 토글스위치를 활성화하지 않은 경우입니다.
* **해결:** 스위치를 켜고 Colab 우측 상단의 '다시 연결' 혹은 '세션 다시 시작'을 누르게 지도합니다.


2. **`RateLimitError: You exceeded your current quota...`**
* **원인:** 개인 OpenAI 계정의 무료 크레딧이 만료되었거나, 카드 결제 연동(Billing) 및 선불 충전 누락으로 잔액이 $0인 상태입니다.
* **💡 단골 질문 대응:** 수강생이 "저 매달 20달러씩 내는 ChatGPT Plus 구독자인데 왜 에러가 나죠?"라고 묻는다면, "💡 **ChatGPT Plus 구독료와 개발용 API 사용 요금은 정산 체계가 완전히 별개입니다!**"라고 안내한 후 최소 $5 선불 충전을 안내하거나 강사의 백업용 키를 배포합니다.


3. **`JSONDecodeError: Expecting value: line 1 column 1 (char 0)`**
* **원인:** `response_format={"type": "json_object"}` 모드를 사용하면서 `system` 프롬프트에 "반드시 JSON 구조로 출력하라"는 텍스트 명시를 누락하여 OpenAI 자체 필터 에러가 발생한 경우입니다.
* **해결:** 반드시 `system` 메시지에 JSON 형태로 출력하라는 명확한 명령을 덧붙여야 함을 강조해 줍니다.
