> **"자주 반복하는 조리 과정을 버튼 하나로 자동화하라!"**

> 매번 재료를 썰고, 불 조절을 하고, 양념을 치는 과정을 일일이 코딩하면 주방이 금방 엉망이 됩니다.
> `def`는 **재료(매개변수)** 만 넣으면 알아서 요리가 완성되어 나오는 **"만능 자동 조리기(에어프라이어, 믹서기)"** 를 우리가 직접 조립하는 마법입니다.
:

### 🍳 Basic 실습: 만능 양념 믹서기 조립하기





````python
# 믹서기 설계도(def) 작성하기
def mix_sauce(sauce_name):
    # [주방 절대 규칙: 스코프(Scope)]
    # 믹서기 안에서 일어난 비밀 배합(secret_recipe)은 외부 주방에서는 절대 볼 수 없습니다!
    secret_recipe = "비법 마늘 가루"

    sauce_name = sauce_name + " 소스"
    return f"✨ [완성] {sauce_name}가 맛있게 제조되었습니다! 🍯" # return은 완성된 요리 서빙!

# 믹서기 버튼 누르기 (함수 호출)
print(mix_sauce("간장"))
print(mix_sauce("매콤"))

# 만약 주방에서 믹서기 내부 변수인 secret_recipe를 강제로 보려고 하면?
# print(secret_recipe) ➡️ 에러 발생! "함수 내부의 비밀은 밖에서 들리지 않아요!"
````

````python
# ✅ 미션챌린지 1. 자동 오더 접수기
# 🔹 문제: 메뉴 이름을 넣으면 "○○ 주문 접수되었습니다! 🛎️"를 반환하는 함수를 만드세요.
# 🔹 출력 예시: 스테이크 주문 접수되었습니다! 🛎️
def accept_order(menu):
    return f"{menu} 주문 접수되었습니다! 🛎️"

print(accept_order("스테이크"))


# ✅ 미션챌린지 2. 만능 주스 착즙기
# 🔹 문제: 과일 이름을 넣으면 "○○ 주스 착즙 완료! 🧃"를 반환하는 함수를 만드세요.
def make_juice(fruit):
    return f"{fruit} 주스 착즙 완료! 🧃"

print(make_juice("사과"))


# ✅ 미션챌린지 3. 오븐 온도 자동 안심 경고기
# 🔹 문제: 온도를 넣었을 때 200도 이상이면 "위험! 불 낮추세요! 🔥", 아니면 "안전 조리 중 ♨️"을 반환하세요.
def check_oven(temp):
    if temp >= 200:
        return "위험! 불 낮추세요! 🔥"
    else:
        return "안전 조리 중 ♨️"

print(check_oven(210))
print(check_oven(180))


# ✅ 미션챌린지 4. 자동 고기 굽기 등급 판독기
# 🔹 문제: 스테이크 굽기 상태를 넣으면 칭찬 멘트를 돌려주는 함수를 만드세요.
def steak_grade(state):
    if state == "미디움":
        state = "육즙 가득 " + state
    elif state == "웰던":
        state = "바짝 익힌 " + state
    else:
        state = " 부드러운 " + state
    return f"와! {state} 스테이크 너무 맛있어 보여요! 🥩"

print(steak_grade("미디움"))


# ✅ 미션챌린지 5. 전 직원 단체 서빙 자동화
# 🔹 문제: 반복문(for)과 함수를 조합하여, 모든 서빙 알바생에게 명령을 전파하세요.
def serve_food(waiter_name):
    return f"셰프 알림: {waiter_name} 매니저님, 서빙 출발해 주세요! 🏃‍♂️"

crew = ["민지", "하니", "해린", "혜인"]
for member in crew:
    print(serve_food(member))
````
