#### 📦 데이터 분석 주방의 4대 핵심 수납함

| 수납함 키워드 | 주방에서의 진짜 의미 👨‍🍳 | NumPy / 데이터 분석 연계 포인트 🧠 |
| --- | --- | --- |
| **`List` (리스트)** | **재료 순서 바구니**  (재료를 순서대로 줄 세워 담기) | **NumPy Array(배열)**로 진화!  대량의 데이터를 빛의 속도로 계산할 때 기반이 됨 |

| **`Dictionary` (딕셔너리)** | **견출지 붙은 양념통 선반**  (`이름:내용물` 짝꿍 관리) | **Pandas DataFrame(표 데이터)**으로 진화!  `열 이름:데이터` 구조의 핵심 뼈대 |
| **`Set` (셋)** | **똑똑한 재료 분류함**  (중복 재료는 제거하고 종류만 남기기) | 데이터셋에서 **고유값(Unique Value)** 컬럼 종류나 카테고리를 뽑아낼 때 사용 |

#### 🧺 1. 재료 순서 바구니: `List` (리스트)

> **"순서대로 차곡차곡 담고, 언제든 재료를 넣었다 뺐다 하라!"**

### 🍳 Basic 실습: 기본 주방 바구니 다듬기





````python
# 💡 [1단계] 재료 바구니 만들기 (생성)
# 대괄호 []를 사용하여 재료를 순서대로 담습니다.
basket = ["당근", "양파", "고기"]
print("🧺 현재 재료 바구니:", basket)

# ➕ [2단계] 새로운 재료 추가하기 (append)
# 손님이 추가 주문을 해서 '대파'를 바구니 맨 뒤에 넣습니다.
basket.append("대파")
print("➕ 대파 추가 후:", basket)

# ✏️ [3단계] 재료 교체하기 (수정)
# 고기 대신 '버섯'으로 요리해 달라고 하시네요! (파이썬 번호표는 0번부터 시작!)
# ["당근", "양파", "고기", "대파"] -> 고기는 2번 자리에 있습니다.
basket[2] = "버섯"
print("✏️ 재료 수정 후:", basket)

# 🗑️ [4단계] 상한 재료 꺼내기 (삭제)
# 요리에 쓰지 않을 '양파'를 바구니에서 꺼냅니다.
basket.remove("양파")
print("🗑️ 양파 제거 후:", basket)

# 🔍 [5단계] 창고에 재료가 있는지 확인 (검색)
# 바구니에 '버섯'이 들어있는지 확인하고 요리를 시작합니다.
if "버섯" in basket:
    print("🍳 찾았다! 버섯이 바구니에 있으니 요리를 시작합니다!")
else:
    print("😢 버섯이 없어서 마트로 장 보러 갑니다.")
````

###### 🧠 NumPy 주방 연결고리 (리스트의 한계)

> 파이썬 리스트 바구니는 아무 재료(`"문자"`, `123`, `True`)나 다 담을 수 있어서 편리하지만, 1,000만 개쯤 되는 대량의 고기를 썰 때는 속도가 느려집니다.
> 나중에 배울 **NumPy 고속 조리대(Array)** 는 **"오직 한 종류의 데이터만"**  담는 대신, 속도를 100배 이상 빠르게 업그레이드한 특수 바구니입니다!

#### 📅 2. 변하지 않는 비밀 레시피: `Tuple` (튜플)

> **"절대로 수정 불가능! 주방의 안전을 책임지는 고정 값!"**

### 🍳 Basic 실습: 고정 요일 및 메뉴 고정하기




````python
# 🧁 [1단계] 고정 레시피 만들기 (생성)
# 소괄호 ()를 사용하며, 한 번 정하면 절대 주방장이 수정할 수 없습니다.
fixed_menu = ("비빔밥", "불고기", "냉면")
print("📅 이번 주 고정 메뉴:", fixed_menu)

# ✏️ [2단계] 강제로 메뉴를 바꾸려고 하면? (수정 시도)
# 알바생이 실수로 불고기를 '피자'로 바꾸려고 하면 주방에서 경고(에러)가 발생합니다!
try:
    fixed_menu[1] = "피자"
except TypeError:
    print("🚨 경고: 튜플 레시피는 절대 수정할 수 없습니다! (수정 불가 봉인)")

# 👀 [3단계] 순서대로 꺼내서 손님에게 보여주기 (인덱싱)
print("첫 번째 시그니처 메뉴:", fixed_menu[0])

# 🎲 [4단계] 등록된 고정 메뉴는 몇 개? (길이 조회)
print("고정 메뉴 총 개수:", len(fixed_menu))
````

##### 🧠 NumPy 주방 연결고리 (튜플의 역할)

> 데이터 분석을 할 때, 수천 행의 데이터 표를 가져오게 됩니다. 이때 이 표가 **"몇 행 몇 열"** 로 이루어져 있는지 알려주는 크기 데이터(`shape`)가 바로 **튜플** 형태(`(10000, 5)`)로 반환됩니다. 데이터 크기가 분석 도중에 마음대로 바뀌면 대형 사고가 나기 때문에, 안전하게 고정하기 위해 튜플을 씁니다!

#### 🏷️ 3. 견출지 붙은 양념통 선반: `Dictionary` (딕셔너리)

> **"이름표(Key)를 부르면 내용물(Value)이 튀어나온다!"**

##### 🍳 Basic 실습: 주방 재료 가격표 매니저





````python
# 🎁 [1단계] 주방 재료 가격표 만들기 (생성)
# 중괄호 {} 안에 '이름표: 내용물' 형태로 짝을 지어줍니다.
menu_prices = {"비빔밥": 9000, "불고기": 12000, "냉면": 8000}
print("📝 주방 메뉴판:", menu_prices)

# ✏️ [2단계] 물가 상승! 가격 수정하기 (수정)
# 냉면 가격을 9,000원으로 올립니다.
menu_prices["냉면"] = 9000
print("✏️ 가격 인상 후:", menu_prices)

# ➕ [3단계] 신메뉴 추가하기 (추가)
# 새로운 메뉴인 '만두'를 가격표에 등록합니다.
menu_prices["만두"] = 5000
print("➕ 신메뉴 추가 후:", menu_prices)

# 🗑️ [4단계] 단종된 메뉴 삭제하기 (삭제)
# 이제 판매하지 않는 '비빔밥'을 메뉴판에서 지웁니다.
del menu_prices["비빔밥"]
print("🗑️ 비빔밥 단종 후:", menu_prices)

# 🔍 [5단계] 손님이 주문한 메뉴 가격 바로 조회 (검색)
# 이름표만 부르면 가격이 바로 튀어나옵니다.
print("불고기는 얼마인가요?", menu_prices["불고기"], "원입니다! 💵")
````

##### 🧠 NumPy/Pandas 주방 연결고리 (딕셔너리의 변신)

> 데이터 분석의 핵심 도구인 **Pandas DataFrame(표 데이터)**을 만들 때 이 딕셔너리가 주인공이 됩니다.
> `{"이름": ["철수", "영희"], "점수": [90, 80]}` 처럼 딕셔너리를 짜서 넣어주면, 이름표(`Key`)가 **엑셀의 열 이름(Column)**이 되고 내용물(`Value`)이 **그 열의 데이터**가 되는 마법이 펼쳐집니다!

#### 🧽 4. 똑똑한 재료 분류함: `Set` (셋)

> **"중복은 허용하지 않는다! 오직 재료의 순수한 종류만 남겨라!"**

##### 🍳 Basic 실습: 오늘 들어온 주방 재료 종류 확인하기


````python
# 🧃 [1단계] 재료 분류함 만들기 (중복 제거 자동화)
# 중괄호 {}를 사용하며, 똑같은 재료가 여러 번 들어와도 하나만 남깁니다.
ordered_ingredients = {"당근", "양파", "당근", "고기", "양파"}
print("📦 오늘 준비된 고유 재료 종류:", ordered_ingredients) # 중복이 사라져 있어요!

# ➕ [2단계] 새로운 재료 채워넣기 (add)
# 분류함에 '대파'를 추가합니다. (이미 있는 재료를 또 넣어도 변화가 없습니다.)
ordered_ingredients.add("대파")
ordered_ingredients.add("당근") # 이미 있어서 추가 안 됨
print("➕ 재료 추가 후 분류함:", ordered_ingredients)

# 🤝 [3단계] 오전반 셰프와 오후반 셰프의 재료 비교 (집합 연산)
chef_A = {"당근", "양파", "고기"}
chef_B = {"고기", "대파", "마늘"}

# 1) 교집합 (&) : 두 셰프가 공통으로 가지고 있는 재료
print("🤝 공통 재료 (교집합):", chef_A & chef_B)

# 2) 합집합 (|) : 주방에 있는 모든 재료 종류
print("🚀 전체 재료 (합집합):", chef_A | chef_B)

# 3) 차집합 (-) : A 셰프만 가지고 있는 독점 재료
print("🔥 A 셰프 독점 재료 (차집합):", chef_A - chef_B)
````

##### 🧠 데이터 분석 주방 연결고리 (셋의 가치)

> 설문조사 데이터나 고객 구매 이력 100만 건을 분석할 때, 거주지 컬럼에 `서울, 서울, 부산, 대구, 서울...` 처럼 수많은 중복 데이터가 들어있습니다.
> 이때 **"우리 고객들이 사는 지역의 '종류'가 총 몇 개지?"**를 알고 싶을 때 셋(`Set`)의 원리를 이용한 `.unique()` 함수를 써서 중복을 한 방에 청소합니다!

##### 🔥 요약 코닝 페이퍼

* **`List` `[]`** : 넣는 대로 순서 유지, 언제든 수정 가능 (**가장 많이 씀!**)
* **`Tuple` `()`** : 수정 절대 불가, 주방 안전 보관용 (**차원, 크기 지정**)
* **`Dict` `{K:V}`** : 이름표-내용물 세트 (**엑셀 표 데이터의 뼈대**)
* **`Set` `{}`** : 중복은 자동 삭제 (**고유값, 종류 분류기**)

````python
# ===================================================================
# 👨‍🍳 [1교시 미션] 대형 레스토랑의 '신선실 재료 바구니' 관리하기
# ===================================================================

# ✅ 미션챌린지 1. 오늘 사용할 핵심 재료 바구니 만들기
# 🔹 문제: 아래 리스트에서 첫 번째(인덱스 0번) 재료를 출력하세요.
# 🔹 출력 예시: 삼겹살
fresh_basket = ["삼겹살", "대파", "양파"]
print(fresh_basket[0])


# ✅ 미션챌린지 2. 깜빡한 재료 추가하기
# 🔹 문제: 바구니 맨 뒤에 '통마늘'을 추가하고 바구니 전체를 출력하세요.
# 🔹 출력 예시: ['삼겹살', '대파', '양파', '통마늘']
fresh_basket.append("통마늘")
print(fresh_basket)


# ✅ 미션챌린지 3. 거래처 사정으로 재료 긴급 교체 (수정)
# 🔹 문제: 두 번째 재료(대파)를 '쪽파'로 바꾸고 출력하세요.
# 🔹 출력 예시: ['삼겹살', '쪽파', '양파', '통마늘']
fresh_basket[1] = "쪽파"
print(fresh_basket)


# ✅ 미션챌린지 4. 메뉴 폐기로 인한 재료 제외 (삭제)
# 🔹 문제: 바구니에서 '양파'를 삭제하고 출력하세요.
# 🔹 출력 예시: ['삼겹살', '쪽파', '통마늘']
fresh_basket.remove("양파")
print(fresh_basket)


# ✅ 미션챌린지 5. 재료 창고 재고 확인 (검색)
# 🔹 문제: 바구니에 '통마늘'이 있으면 "재료 준비 완료! 🧄"를 출력하세요.
# 🔹 출력 예시: 재료 준비 완료! 🧄
if "통마늘" in fresh_basket:
    print("재료 준비 완료! 🧄")
else:
    print("재료가 부족해요 😢")
````

### 📅 2교시: 튜플(Tuple) 미션챌린지 리모델링

> **기존:** 생일 달력, 교통편, 요일 시간표
> **변경:** **[레스토랑 고정 정찰제 메뉴판]** ➡️ 데이터 분석에서 변하지 않는 '데이터의 크기/차원(Shape)'으로 가기 위한 빌드업

```python
# ===================================================================
```

````python

# ✅ 미션챌린지 1. 재료별 칼로리 등록하기
# 🔹 문제: 이름표(Key)는 재료명, 정보(Value)는 칼로리(숫자)로 딕셔너리를 만드세요.
# 🔹 출력 예시: {'아보카도': 160, '닭가슴살': 165, '고구마': 86}
calorie_table = {'아보카도': 160, '닭가슴살': 165, '고구마': 86}
print(calorie_table)


# ✅ 미션챌린지 2. 영양 성분 조사 후 칼로리 대폭 수정
# 🔹 문제: '닭가슴살'의 칼로리를 120으로 수정해 보세요.
# 🔹 출력 예시: {'아보카도': 160, '닭가슴살': 120, '고구마': 86}
calorie_table['닭가슴살'] = 120
print(calorie_table)


# ✅ 미션챌린지 3. 새로운 다이어트 재료 추가하기
# 🔹 문제: '샐러드'를 추가하고 칼로리는 20으로 넣어보세요.
# 🔹 출력 예시: {'아보카도': 160, '닭가슴살': 120, '고구마': 86, '샐러드': 20}
calorie_table['샐러드'] = 20
print(calorie_table)


# ✅ 미션챌린지 4. 품절된 재료 삭제하기
# 🔹 문제: '고구마'를 딕셔너리에서 완전히 지워보세요.
# 🔹 출력 예시: {'아보카도': 160, '닭가슴살': 120, '샐러드': 20}
del calorie_table['고구마']
print(calorie_table)


# ✅ 미션챌린지 5. 특정 재료의 칼로리만 쏙 꺼내기
# 🔹 문제: 이름표를 이용해 '아보카도'의 칼로리만 출력해 보세요.
# 🔹 출력 예시: 아보카도의 칼로리는 160 kcal 입니다! 🥑
print(f"아보카도의 칼로리는 {calorie_table['아보카도']} kcal 입니다! 🥑")
````

### 🧽 4교시: 셋(Set) 미션챌린지 리모델링

> **기존:** 캐릭터 스티커, 좋아하는 운동
> **변경:** **[오전/오후 파트타임 셰프들의 보유 재료 비교]** ➡️ 데이터 분석에서 중복 데이터를 청소하는 '.unique()' 및 집합 연산 빌드업

```python
# ===================================================================
# 👨‍🍳 [4교시 미션] 중복 재료 제거 및 A/B 셰프의 재료고 분석하기
# ===================================================================

# ✅ 미션챌린지 1. 중복 가득한 주문서에서 '순수 재료 종류'만 남기기
# 🔹 문제: 중복된 재료들이 섞인 리스트를 셋으로 변환하여 중복을 자동 청소하세요.
# 🔹 출력 예시: {'새우', '토마토', '치즈'} (순서는 바뀔 수 있습니다)
duplicate_orders = ["새우", "토마토", "새우", "치즈", "토마토"]
clean_ingredients = set(duplicate_orders)



# ✅ 미션챌린지 2. 분류함에 새로운 재료 추가하기
# 🔹 문제: 셋에 '바질'을 추가해 보세요. (단, 여러 번 추가해도 한 번만 들어가는지 확인!)
clean_ingredients.add("바질")
clean_ingredients.add("바질") # 중복 추가 시도



# ✅ 미션챌린지 3. 안 쓰는 재료 분류함에서 빼기
# 🔹 문제: 셋에서 '토마토'를 제거하세요.



# ✅ 미션챌린지 4. 두 셰프가 '공통으로' 가지고 있는 최고 인기 재료 찾기 (교집합)
# 🔹 문제: 두 셰프의 재료 셋에서 겹치는 종목(&)을 출력해보세요.
# 🔹 출력 예시: {'치즈'}
chef_gold = {'새우', '치즈', '바질'}
chef_silver = {'감자', '스테이크', '치즈'}




# ✅ 미션챌린지 5. 두 셰프의 재료를 합치면 총 몇 종류의 요리를 할 수 있을까? (합집합)
# 🔹 문제: 두 셰프의 재료를 중복 없이 모두 합친(|) 결과를 출력하세요.
# 🔹 출력 예시: {'새우', '치즈', '바질', '감자', '스테이크'}


```


````python

# ✅ 미션챌린지 1. 중복 가득한 주문서에서 '순수 재료 종류'만 남기기
# 🔹 문제: 중복된 재료들이 섞인 리스트를 셋으로 변환하여 중복을 자동 청소하세요.
# 🔹 출력 예시: {'새우', '토마토', '치즈'} (순서는 바뀔 수 있습니다)
duplicate_orders = ["새우", "토마토", "새우", "치즈", "토마토"]
clean_ingredients = set(duplicate_orders)
print("순수 재료 종류 ->", clean_ingredients)


# ✅ 미션챌린지 2. 분류함에 새로운 재료 추가하기
# 🔹 문제: 셋에 '바질'을 추가해 보세요. (단, 여러 번 추가해도 한 번만 들어가는지 확인!)
clean_ingredients.add("바질")
clean_ingredients.add("바질") # 중복 추가 시도
print("새로운 재료 add ->", clean_ingredients)


# ✅ 미션챌린지 3. 안 쓰는 재료 분류함에서 빼기
# 🔹 문제: 셋에서 '토마토'를 제거하세요.
clean_ingredients.remove('토마토')
print("재료 제거 remove ->", clean_ingredients)


# ✅ 미션챌린지 4. 두 셰프가 '공통으로' 가지고 있는 최고 인기 재료 찾기 (교집합)
# 🔹 문제: 두 셰프의 재료 셋에서 겹치는 종목(&)을 출력해보세요.
# 🔹 출력 예시: {'치즈'}
chef_gold = {'새우', '치즈', '바질'}
chef_silver = {'감자', '스테이크', '치즈'}

common_items = chef_gold & chef_silver
print('두 셰프가 공통으로 가진 재료 (교집합) ->', common_items)


# ✅ 미션챌린지 5. 두 셰프의 재료를 합치면 총 몇 종류의 요리를 할 수 있을까? (합집합)
# 🔹 문제: 두 셰프의 재료를 중복 없이 모두 합친(|) 결과를 출력하세요.
# 🔹 출력 예시: {'새우', '치즈', '바질', '감자', '스테이크'}
all_items = chef_gold | chef_silver
print('주방에 있는 모든 재료 종류 (합집합) ->', all_items)
````
