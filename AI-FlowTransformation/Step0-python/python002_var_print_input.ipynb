#### 2. 셰프의 도구상자: **변수 (Variable)**

변수는 데이터를 저장하는 메모리 공간의 '이름표'입니다. 파이썬에서는 변수에 값을 할당하면 자동으로 변수가 생성되고, 필요할 때 이 이름표를 사용하여 저장된 데이터에 접근할 수 있습니다. 변수를 사용하면 코드를 더 유연하고 효율적으로 만들 수 있으며, 특히 반복되는 값을 관리할 때 매우 유용합니다.

*   `nickname = "슈퍼코딩마스터3000"`은 `"슈퍼코딩마스터3000"`이라는 문자열을 `nickname`이라는 변수에 저장합니다.
*   `age = 30`은 숫자 `30`을 `age`라는 변수에 저장합니다.
*   `print(f"제 나이는 {age}세입니다.")`와 같이 변수를 `print()` 함수와 함께 사용하여 저장된 값을 출력할 수 있습니다.

```python

# 변수 사용 및 출력

print(f"나이: {my_age}세")

```

#### 3. 셰프의 재료 상자: **자료형 (Data Types)**

파이썬에서는 데이터의 종류에 따라 '자료형'이 구분됩니다. 각 자료형은 데이터가 어떻게 저장되고 처리되는지를 결정합니다. 올바른 자료형을 이해하고 사용하는 것은 효율적인 코드를 작성하는 데 중요합니다.

**왜 자료형이 필요할까요?**

상상해보세요. 주방에서 "사과 3개"라는 메모를 남겼을 때, 우리는 3이 '수량'을 의미한다는 것을 자연스럽게 이해합니다. 하지만 컴퓨터는 그저 '3'이라는 글자를 숫자로 처리해야 할지, 아니면 '3'이라는 글자 자체로 처리해야 할지 알 수 없습니다. 이때 자료형은 컴퓨터에게 "3은 숫자로 계산해!" 또는 "'3'은 글자로 다른 글자와 합쳐줘!"와 같은 명확한 지시를 내려주는 역할을 합니다. 이처럼 자료형을 통해 데이터의 의미와 사용 방식을 명확히 하여 컴퓨터가 오류 없이 정확한 연산을 수행할 수 있도록 합니다.

| 자료형  | 예시 값         | 설명                    |
| ------- | --------------- | ----------------------- |
| `int`   | `10`, `-3`      | 정수형 숫자             |
| `float` | `3.14`, `-0.5`  | 소수 포함된 실수형 숫자 |
| `str`   | `"Hello"`       | 문자열 자료형           |
| `bool`  | `True`, `False` | 논리값 (참/거짓)        |

*   변수는 **데이터를 저장할 이름표**예요.     예시: `name = "Sally"`
*   할당(assignment):  `변수명 = 값`
*   문자열 결합: `"파이썬" + " 최고"` → `"파이썬 최고"`

#### 4. 셰프와 손님들의 소통 창구: `input()` 함수

`input()` 함수는 사용자로부터 직접 데이터를 입력받을 때 사용됩니다. 이 함수는 사용자에게 메시지를 보여주고, 사용자가 키보드로 입력한 내용을 읽어와 **항상 문자열(str) 형태로 반환**합니다.

주방에서 손님에게 주문을 받을 때와 같아요. 손님이 "아메리카노"라고 말하면 그 자체로 문자열이지, 숫자가 아니죠? `input()` 함수도 마찬가지입니다. 따라서 숫자를 입력받아 계산해야 할 경우, `int()`나 `float()` 같은 자료형 변환 함수를 사용하여 숫자로 바꿔주는 과정이 필수적입니다.

```python
# 사용자에게 이름을 물어보고, 입력받은 이름을 name 변수에 저장합니다.
name = input("당신의 이름은 무엇인가요? ")
print(f"환영합니다, {name}님!")

# 사용자에게 나이를 물어보고, 입력받은 나이를 age_str 변수에 저장합니다.
# input() 함수는 항상 문자열을 반환하므로, age_str은 문자열입니다.
age_str = input("당신의 나이는 몇 살인가요? ")

# 문자열인 age_str을 정수(int)로 변환하여 age_int 변수에 저장합니다.
age_int = int(age_str)

# 정수형 나이를 사용하여 계산을 수행합니다.
next_year_age = age_int + 1
print(f"내년에는 {next_year_age}살이 되시겠군요!")
```

````python
# 사용자에게 이름을 물어보고, 입력받은 이름을 name 변수에 저장합니다.
name = input("당신의 이름은 무엇인가요? ")
print(f"환영합니다, {name}님!")

# 사용자에게 나이를 물어보고, 입력받은 나이를 age_str 변수에 저장합니다.
# input() 함수는 항상 문자열을 반환하므로, age_str은 문자열입니다.
age_str = input("당신의 나이는 몇 살인가요? ")

# 문자열인 age_str을 정수(int)로 변환하여 age_int 변수에 저장합니다.
age_int = int(age_str)

# 정수형 나이를 사용하여 계산을 수행합니다.
next_year_age = age_int + 1
print(f"내년에는 {next_year_age}살이 되시겠군요!")
````

#### 5. 셰프의 마법 변신술: 자료형 변환 (Type Conversion)

데이터는 원래의 자료형 외에 다른 자료형으로 '변환'될 수 있습니다. 이를 자료형 변환(Type Conversion)이라고 합니다. 특히 `input()` 함수로 사용자 입력을 받을 경우, 입력된 모든 내용은 문자열(`str`)로 처리되기 때문에 숫자로 계산하려면 반드시 자료형 변환이 필요합니다. 셰프가 재료를 요리에 맞게 다듬는 것처럼, 파이썬도 데이터가 필요한 형태로 변환되어야 올바른 연산을 수행할 수 있습니다.

`int()`, `float()`, `str()` 등의 함수를 사용하여 자료형을 변환할 수 있습니다.

```python
# 일반적인 문제 상황: 사용자로부터 두 숫자를 입력받아 더하려고 합니다.
# input() 함수는 항상 문자열을 반환합니다.
num1_str = input("첫 번째 숫자를 입력하세요: ") # 예: "10"
num2_str = input("두 번째 숫자를 입력하세요: ") # 예: "20"

# 현재 상태로 더하면 문자열 결합이 일어납니다.
print(f"문자열로 더한 결과: {num1_str + num2_str}") # 결과: "1020"

# 해결책: int() 함수를 사용하여 문자열을 정수로 변환합니다.
num1_int = int(num1_str)
num2_int = int(num2_str)

# 이제 올바른 덧셈 연산이 가능합니다.
print(f"정수로 변환하여 더한 결과: {num1_int + num2_int}") # 결과: 30

# 다른 자료형 변환 예시:
# 정수를 실수로 변환
int_val = 10
float_val = float(int_val)
print(f"정수 {int_val}을 실수로 변환: {float_val} (자료형: {type(float_val)})")
```

#### Q2: 간단한 주문 계산 프로그램 만들기

**목표:** `input()` 함수를 사용하여 손님으로부터 주문 수량을 입력받고, 자료형 변환을 통해 총 가격을 계산하는 프로그램을 만들어 봅시다.


**예시 입력 및 출력:**

```
첫 번째 숫자를 입력하세요: 5
두 번째 숫자를 입력하세요: 3
입력하신 두 숫자의 합은 8입니다.
```

````python
# 사용자에게 첫 번째 숫자를 입력받아 문자열로 저장합니다.
number1_str = input("첫 번째 숫자를 입력하세요: ")

# 사용자에게 두 번째 숫자를 입력받아 문자열로 저장합니다.
number2_str = input("두 번째 숫자를 입력하세요: ")

# 입력받은 문자열을 정수(int)로 변환합니다.
number1_int = int(number1_str)
number2_int = int(number2_str)

# 두 정수를 더하여 결과를 계산합니다.
sum_of_numbers = number1_int + number2_int

# 결과를 출력합니다.
print(f"입력하신 두 숫자의 합은 {sum_of_numbers}입니다.")
````
