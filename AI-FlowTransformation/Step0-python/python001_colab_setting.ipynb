#### 1.  colab 사용 환경 설정

>Google Colab은 별도의 복잡한 설치 과정 없이 웹 브라우저에서 바로 사용할 수 있는 개발 환경을 제공합니다.

1. **Google 계정 로그인 및 접속:** [Google Colab](https://colab.research.google.com/)에 접속하여 Google 계정으로 로그인합니다.
2. **자동 저장 기능 활용:** Colab에서 새 노트북을 생성하거나 기존 파일을 열어 작업하면, 해당 파일은 사용자의 **Google 드라이브 내 'Colab Notebooks' 폴더**에 자동으로 저장됩니다.
* **실시간 동기화:** 별도의 저장 버튼을 누르지 않아도 드라이브와 실시간으로 연동되어 작업 내용이 안전하게 반영됩니다.
