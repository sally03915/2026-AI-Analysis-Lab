> **"창고 문을 열고(`open`), 재료를 입고하거나(`'w'`), 보관된 재료를 확인하라(`'r'`)!"**
> 파이썬 코드 창에 `basket = [...]`라고 직접 치는 것은 매번 마트에서 재료를 소량으로 사 오는 것과 같습니다. 프로그램이 종료되면 다 사라지죠.
> `File` 입출력은 **외부 식재료 창고(텍스트 파일, 엑셀 파일, CSV 파일)에 대량의 데이터를 보관하거나 가져오는 작업**입니다.

* **`'w'` (Write, 쓰기 모드):** 창고에 새 재료 기록하기 (기존 기록은 지워지고 새로 써짐)
* **`'r'` (Read, 읽기 모드):** 창고에 보관된 기록 읽어오기
* **`with open(...)`:** 재료를 꺼낸 뒤 **창고 문(뚜껑)을 알아서 닫아주는** 안전 고리

<br/>
<br/>  

> 👨‍🍳 **셰프의 비밀 노트:**
> 실습하기 전 미리 파일을 컴퓨터에 만들 필요 없어요! `'w'`(쓰기) 명령어를 실행하면, 파이썬이 알아서 새 텍스트 파일을 자동으로 뚝딱 만들어 줍니다.

* **6-1 ('w' 모드): 오늘 밤 만든 특제 소스 레시피 창고에 저장하기**


> 👨‍🍳 **셰프의 특급 힌트:**
> 실습 전 미리 파일을 만들 필요 없어요! `'w'`(쓰기) 명령어가 포함된 코드를 실행하면,
> 파이썬이 코랩 창고에 알아서 새 텍스트 파일을 자동으로 뚝딱 만들어 줍니다.
> 코드를 실행한 후 왼쪽 폴더(📁) 아이콘을 열어 파일이 생겼는지 꼭 확인해 보세요!


````python
# 1. 'secret.txt'라는 파일을 'w'(쓰기) 모드로 열어(open) f라고 부르겠습니다.
with open('secret.txt', 'w', encoding='utf-8') as f:
    f.write("비법 소스: 간장 3숟갈, 올리고당 1숟갈 🍯") # 파일에 내용 적기

print("📝 창고 파일에 레시피 저장 완료!")
````

<br/>
<br/>

* **6-2 ('r' 모드): 다음 날 아침, 창고에서 레시피 읽어오기**

````python
# 2. 방금 만든 'secret.txt' 파일을 'r'(읽기) 모드로 열어 f라고 부르겠습니다.
with open('secret.txt', 'r', encoding='utf-8') as f:
    recipe_content = f.read() # 파일 내용 전부 읽어와서 변수에 담기
    print("📖 창고에서 읽어온 내용:", recipe_content)
````

````python
# ===================================================================
# 📋 [미션 1] 파일명: kitchen_log.txt  | 내용: "오늘 주방 마감 이상 무! 깨끗하게 청소 완료."
# ===================================================================
# ✅ 미션챌린지 1. 주방 일지 마감 기록하기
# 🔹 문제: 'kitchen_log.txt' 파일에 마감 한 줄 평을 적고 다시 읽어서 출력하세요.
with open('kitchen_log.txt', 'w', encoding='utf-8') as f:
    f.write("오늘 주방 마감 이상 무! 깨끗하게 청소 완료.")

with open('kitchen_log.txt', 'r', encoding='utf-8') as f:
    print(f.read())


# ===================================================================
# 📋 [미션 2] 파일명: vegetables.txt   | 내용: "당근 5개, 양파 10개 남음 🧅"
# ===================================================================
# ✅ 미션챌린지 2. 오늘 밤 남은 야채 재고 파일 저장
# 🔹 문제: 'vegetables.txt' 파일에 재고를 적고 다시 읽어보세요.
with open('vegetables.txt', 'w', encoding='utf-8') as f:
    f.write("당근 5개, 양파 10개 남음 🧅")

with open('vegetables.txt', 'r', encoding='utf-8') as f:
    print(f.read())


# ===================================================================
# 📋 [미션 3] 파일명: meat_delivery.txt | 내용: 거래처별 고기 목록 (줄바꿈 \n 활용)
# ===================================================================
# ✅ 미션챌린지 3. 거래처별 고기 입고 목록 작성 (\n 활용 줄바꿈)
# 🔹 문제: 'meat_delivery.txt' 파일에 여러 종류의 고기를 줄바꿈하여 적고 읽어보세요.
with open('meat_delivery.txt', 'w', encoding='utf-8') as f:
    f.write("A거래처: 소고기 등심 20kg\nB거래처: 돼지고기 삼겹살 50kg")

with open('meat_delivery.txt', 'r', encoding='utf-8') as f:
    print(f.read())


# ===================================================================
# 📋 [미션 4] 파일명: secret_sauce.txt  | 내용: "비법 소스 비율 = 간장 3 : 설탕 1 : 참기름 1 🍯"
# ===================================================================
# ✅ 미션챌린지 4. 특별 소스 비밀 레시피 금고 보관
# 🔹 문제: 'secret_sauce.txt'에 비법을 적고 읽어보세요.
with open('secret_sauce.txt', 'w', encoding='utf-8') as f:
    f.write("비법 소스 비율 = 간장 3 : 설탕 1 : 참기름 1 🍯")

with open('secret_sauce.txt', 'r', encoding='utf-8') as f:
    print(f.read())


# ===================================================================
# 📋 [미션 5] 파일명: menu_list.txt     | 내용: 판매 메뉴 목록 1~4번 (줄바꿈 \n 활용)
# ===================================================================
# ✅ 미션챌린지 5. 주방 메뉴판 최종본 창고 저장
# 🔹 문제: 판매 메뉴 목록을 여러 줄로 적고 다시 꺼내보세요.
with open('menu_list.txt', 'w', encoding='utf-8') as f:
    f.write("1. 스테이크\n2. 파스타\n3. 리조또\n4. 샐러드")

with open('menu_list.txt', 'r', encoding='utf-8') as f:
    print(f.read())
````
