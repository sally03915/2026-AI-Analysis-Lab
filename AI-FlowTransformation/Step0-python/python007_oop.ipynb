### 👨‍🍳 데이터 분석 주방 : 파이썬 요리 마스터 가이드 6 -   파이썬 기본기 최종장: OOP (객체지향 프로그래밍)

#### 🏢 1교시: 클래스와 객체의 개념 및 구조

> **"매번 주방을 새로 짓지 말고, 완벽한 설계도(`Class`) 하나로 수많은 가맹점 주방(`Object`)을 찍어내라!"**

* **설계도(`Class`) :** 프랜차이즈 본사가 만든 '표준 주방 설계도' (예: 백종원의 빽다방 설계도)
* **가맹점 주방(`Object`) :** 설계도를 보고 실제로 도심에 차린 '실제 매장 주방' (예: 강남점, 홍대점)
* **`__init__()` (생성자) :** 가맹점을 '처음 오픈할 때 세팅해야 할 기본 인테리어'를 지정하는 곳입니다.
* **`self` :** 주방장인 '나 자신'을 가리키는 말입니다. 강남점 주방장도, 홍대점 주방장도 각자 자기 주방을 부를 땐 똑같이 "우리 주방"(`self`)이라고 부르니까요!

```
[설계도: Class] ➡️ [실제 가맹점 주방: Object] ➡️ [주방에서 요리하기: Method]

```

### 🍳 Basic 실습: 만능 주방 설계도 조립하기



````python
class Kitchen:  # 1. 빽다방 주방 설계도(Class) 작성
    def __init__(self, branch_name):  # 2. 처음 가맹점 오픈할 때 간판 이름 달기
        self.branch_name = branch_name

    def welcome(self):  # 3. 주방의 기능(Method) 정의
        print(f"☕ [{self.branch_name}] 오픈했습니다! 맛있는 커피를 만듭니다.")

# 🧪 여기에 실습하세요
shop1 = Kitchen('강남점')  # 2단계 - 설계도 보고 강남점 실제 주방 구축!
shop1.welcome()           # 3단계 - 강남점 주방 가동하기

shop2 = Kitchen('홍대점')  # 2단계 - 설계도 보고 홍대점 실제 주방 구축!
shop2.welcome()           # 3단계 - 홍대점 주방 가동하기
````

````python
# ✅ 미션챌린지 1: 주방 오븐(Oven) 클래스 설계하기
# 🔹 문제: 오븐 설계도를 만들고, 브랜드와 최고 온도를 출력하는 메서드를 정의해보세요.
# 🔹 출력 예시: 이 오븐은 [현대] 브랜드이고, 최고 온도는 180도 입니다.
class Oven:
    def __init__(self, brand, max_temp):
        self.brand = brand
        self.max_temp = max_temp

    def show_info(self):
        print(f"이 오븐은 [{self.brand}] 브랜드이고, 최고 온도는 {self.max_temp}도 입니다.")

# 🧪 여기에 실습하세요
my_oven = Oven("현대", 180)
my_oven.show_info()


# ✅ 미션챌린지 2: 피자 화덕(PizzaOven) 클래스 설계하기
# 🔹 문제: 피자 화덕 설계도를 만들고, 피자 종류와 크기를 접수받는 메서드를 정의해보세요.
# 🔹 출력 예시: 주문하신 [페퍼로니] 피자([라지] 사이즈) 조리를 시작합니다.
class PizzaOven:
    def __init__(self, flavor, size):
        self.flavor = flavor
        self.size = size

    def order(self):
        print(f"주문하신 [{self.flavor}] 피자([{self.size}] 사이즈) 조리를 시작합니다.")

my_pizza = PizzaOven("페퍼로니", "라지")
my_pizza.order()


# ✅ 미션챌린지 3: 요리 로봇(ChefBot) 클래스 설계하기
# 🔹 문제: 주방 보조 요리 로봇 설계도를 만들고 이름과 전담 요리를 소개해 보세요.
# 🔹 출력 예시: 안녕하세요! 저는 요리 로봇 '토비'이고, [청소]를 담당합니다!
class ChefBot:
    def __init__(self, name, job):
        self.name = name
        self.job = job

    def introduce(self):
        print(f"안녕하세요! 저는 요리 로봇 '{self.name}'이고, [{self.job}]를 담당합니다!")

clean_bot = ChefBot("토비", "청소")
clean_bot.introduce()
````

## 👨‍🍳 2교시: 객체 생성 및 인스턴스 변수/메서드 실습


<br/>
<br/>
<br/>

> **"각 가맹점 주방만의 개별 장부(`Variable`)를 관리하고, 일하는 행동(`Method`)을 지시하라!"**

* **인스턴스 변수 :** 각 지점 주방이 개별적으로 가지고 있는 '지점 이름, 남은 예산' 같은 **독립된 장부 정보**입니다.
* **메서드(Method) :** 주방에서 주방장이 부리는 **'요리하기, 청소하기, 매출 계산하기' 같은 실제 행동**입니다.

### 🍳 Basic 실습: 셰프의 하루 가동하기






````python
class Chef:
    def __init__(self, name, age):
        self.name = name  # 인스턴스 변수 (셰프 이름)
        self.age = age    # 인스턴스 변수 (셰프 나이)

    def introduce(self):
        print(f"👨‍🍳 안녕! 나는 {self.name} 셰프, {self.age}살이야!")

    def cook(self):
        print(f"🍳 {self.name} 셰프가 열심히 요리를 하고 있어요!")

    def clean(self):
        print(f"✨ {self.name} 셰프가 주방을 반짝반짝 청소해요!")

# 객체 생성 및 메서드 호출
c1 = Chef('지민', 26)
c1.introduce()
c1.cook()
c1.clean()
````

### 📌 중간 점검 퀴즈 (주방 콘셉트 리모델링)

1. **클래스와 객체의 개념**
* 클래스는 하나의 [ 설계도 ]이고, 객체는 그걸 보고 실제 주방에 구축한 [ 가맹점(매장) ]이에요.


2. **`__init__()`의 역할**
* `__init__`은 객체가 처음 생성될 때 기본 인테리어 세팅을 담당하는 장소예요. 이 함수를 **[ 생성자 ]** 메서드라고 불러요.


3. **`self`의 의미**
* `self`는 다른 지점 주방이 아닌 [ 나 자신(우리 지점) ]의 주방 장부를 가리킬 때 사용해요.


4. **인스턴스 변수와 메서드**
* 인스턴스 변수는 우리 매장의 **[ 장부 정보 ]** 같은 데이터이고, 메서드는 주방에서 할 수 있는 [ 기능(행동) ]이에요.


5. **클래스 사용 순서**
* 클래스(설계도) 만들기 ➡️ **[ 객체(매장) ]** 오픈하기 ➡️ **[ 메서드(기능) ]** 가동하기


````python

# ✅ 미션챌린지 1: 식기세척기 로봇(Robot) – 청소하고 충전하기!
# 🔹 문제: 로봇의 이름을 소개하고, 청소와 전력 충전 행동 메서드를 구현하세요.
class Robot:
    def __init__(self, name):
        self.name = name

    def introduce(self):
        print(f"안녕하세요! 저는 스마트 주방기기 '{self.name}'예요.")

    def clean(self):
        print(f"🧼 {self.name}가 주방 식기를 깨끗하게 세척하고 있어요! ✨")

    def charge(self):
        print(f"🔋 {self.name}가 배터리를 급속 충전 중이에요!")

r1 = Robot('로보')
r1.introduce()
r1.clean()
r1.charge()


# ✅ 미션챌린지 2: 주방 반려견(KitchenPet) – 재료 감시하고 잠자기
# 🔹 문제: 주방 마스코트 동물의 이름과 종류를 말하고, 간식 먹기와 낮잠 기능을 구현하세요.
class KitchenPet:
    def __init__(self, name, kind):
        self.name = name
        self.kind = kind

    def introduce(self):
        print(f"나는 주방 마스코트 {self.kind} '{self.name}'야! 🐱")

    def eat(self):
        print(f"🍚 {self.name}가 남은 식재료 간식을 맛있게 먹어요!")

    def sleep(self):
        print(f"😴 {self.name}가 따뜻한 오븐 옆에서 쿨쿨 자고 있어요...")

pet = KitchenPet("나비", "고양이")
pet.introduce()
pet.eat()
pet.sleep()


# ✅ 미션챌린지 3: 스마트 피자 오븐(PizzaOven) – 오더 넣고 굽기
# 🔹 문제: 피자 종류와 크기를 세팅하고, 주문 접수와 오븐 굽기 행동을 표현해보세요.
class PizzaOven:
    def __init__(self, flavor, size):
        self.flavor = flavor
        self.size = size

    def order(self):
        print(f"🛎️ {self.flavor} 피자 ({self.size} 사이즈) 주문 한 판 들어왔습니다!")

    def bake(self):
        print(f"🔥 {self.flavor} 피자를 오븐에 넣고 노릇하게 굽기 시작합니다!")

p = PizzaOven("페퍼로니", "라지")
p.order()
p.bake()


````

## 🧬 3교시: 상속과 메서드 오버라이딩


<br/>
<br/>
<br/>

> **"기존 주방 설계도를 100% 물려받고(`상속`), 업그레이드할 조리법만 내 스타일로 교체하라(`오버라이딩`)!"**

* **상속 (Inheritance) :** 기본 가맹점 주방 설계도(부모)의 모든 인테리어와 기능을 자식 설계도가 그대로 공짜로 물려받는 마법입니다.
* **메서드 오버라이딩 (Overriding) :** 부모한테 물려받은 조리법이 맘에 안 들거나 신메뉴를 개발하고 싶을 때, **같은 이름의 조리법 메서드를 내 입맛에 맞게 재정의(덮어쓰기)** 하는 핵심 기술입니다.

### 🍳 Basic 실습: 기본 주방 ➡️ 미슐랭 고급 주방으로 업그레이드



````python
class StandardKitchen:  # 부모 클래스: 일반 가맹점 주방
    def make_soup(self):
        print("🥣 평범한 야채 수프를 끓입니다.")

class MichelinKitchen(StandardKitchen):  # 자식 클래스: 상속(StandardKitchen)
    def make_soup(self):  # 오버라이딩: 수프 조리법을 고급화로 개조!
        print("✨ [미슐랭 오버라이딩] 트러플 버섯을 넣은 최고급 수프를 끓입니다! 🍄")

base_shop = StandardKitchen()
base_shop.make_soup()

luxury_shop = MichelinKitchen()
luxury_shop.make_soup()  # 부모 이름과 똑같지만 완전히 업그레이드된 기능이 실행됨!
````

````python
# ✅ 미션챌린지 1: 일반 주방 스태프 ➡️ 수석 주방장 (요리를 훨씬 더 화려하게 합니다)
# 🔹 문제: Chef 스태프 클래스를 상속받아 HeadChef 클래스를 만들고 요리 실력을 업그레이드(오버라이딩)하세요.
class Chef:
    def __init__(self, name):
        self.name = name

    def cook(self):
        print(f"👨‍🍳 {self.name} 스태프가 레시피대로 요리하고 있습니다.")

class HeadChef(Chef):  # 상속
    def cook(self):  # 오버라이딩
        print(f"👑 {self.name} 수석셰프님이 화려한 불쇼를 하며 최고급 요리를 만듭니다! 🔥")

c = Chef("지민")
c.cook()

hc = HeadChef("지민")
hc.cook()


# ✅ 미션챌린지 2: 기본 세척기 ➡️ 초고속 살균 세척기 (설거지 방식의 진화)
# 🔹 문제: SmartDishwasher 클래스를 상속받아 TurboDishwasher 클래스를 만들고 세척 속도를 개조해 보세요.
class SmartDishwasher:
    def __init__(self, name):
        self.name = name
    def clean(self):
        print(f"🧼 {self.name}가 표준 모드로 식기를 세척 중입니다.")

class TurboDishwasher(SmartDishwasher):  # 상속
    def clean(self):  # 오버라이딩
        print(f"⚡ [터보 오버라이딩] {self.name}가 250도 고온 스팀으로 10초 만에 살균 세척 완료! 🧹")

machine1 = SmartDishwasher("식세구형")
machine1.clean()

machine2 = TurboDishwasher("식세신형")
machine2.clean()


# ✅ 미션챌린지 3: 가스 화덕 ➡️ 숲속 참숯 화덕 (굽는 향이 달라져요)
# 🔹 문제: NormalOven 클래스를 상속받아 CharcoalOven 클래스를 만들고, bake 메서드를 새롭게 재정의해보세요.
class NormalOven:
    def __init__(self, flavor):
        self.flavor = flavor
    def bake(self):
        print(f"🔥 {self.flavor} 피자를 가스 불에 굽고 있어요~")

class CharcoalOven(NormalOven):  # 상속
    def bake(self):  # 오버라이딩
        print(f"🪵 [참숯 오버라이딩] {self.flavor} 피자를 300도 참숯 가마에서 향긋하게 굽는 중입니다! 🍕")

oven1 = NormalOven("페퍼로니")
oven1.bake()

oven2 = CharcoalOven("페퍼로니")
oven2.bake()
````
