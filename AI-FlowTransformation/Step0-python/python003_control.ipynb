## 🟨 1. 셰프의 갈림길: `if` / `elif` / `else`

> **"손님의 요구 조건에 따라 다른 요리를 내어라!"**

### 🍳 Basic 실습 & 세부 설명

* **1-1 (if): 가장 단순한 단일 조건문**
* 주문(`order`)이 정확히 `"사탕"`과 일치할 때만 `"좋아!"`라는 메시지를 요리 창에 띄웁니다.


<br/>
<br/>
<br/>


````python
order = "사탕"
if order == "사탕": # 만약 주문이 사탕과 같다면? (주의: 비교할 때는 =를 두 번 씁니다!)
    print("좋아!")
````

<br/>
<br/>
<br/>

* **1-2 (elif / else): 여러 갈래로 나뉘는 조건문**
* 조건은 위에서부터 순서대로 확인합니다. 10점이면 첫 번째 `if`에서 끝나고, 5점이면 `elif`로 내려가며, 둘 다 아니면 `else`(기본값)로 빠집니다.



````python
score = 10
if score == 10:
    print("최고!")     # 10점일 때 실행
elif score == 5:
    print("통과!")     # 10점은 아니지만 5점일 때 실행
else:
    print("노력!")     # 위 조건에 모두 해당하지 않는 나머지 전부
````

### 🚀 Mission Challenge (실습용)

* **Q1. 오븐 온도 조절 시스템**
* **설명:** 사용자에게 온도를 숫자로 입력받아 오븐의 상태를 알려주는 안전 시스템을 만드세요.

## 🟦 2. 셰프의 반복 작업: `for` / `while`

> **"정해진 횟수만큼 재료를 다듬거나, 냄비가 끓을 때까지 계속 저어라!"**

### 🍳 Basic 실습 & 세부 설명

* **2-1 (for 문): 정해진 재료(바구니)에서 하나씩 꺼내기**
* 리스트(재료 바구니) 안에 들어있는 아이템 개수만큼 알아서 차례대로 반복합니다. 바구니가 비면 대기줄이 끝나듯 반복도 자동으로 종료됩니다.



````python
ingredients = ["당근", "양파", "고기"]
for item in ingredients: # 바구니에서 재료를 하나씩 꺼내 'item'이라는 변수에 담음
    print(item)          # 꺼낸 재료의 이름을 출력
````

<br/>
<br/>
<br/>

* **2-2 (for 문 - 유형 B): 셰프의 타이머 `range()`로 원하는 만큼 반복하기**
* 데이터 분석이나 배열(NumPy)을 다룰 때 가장 중요한 도구입니다! 연속된 숫자 주문서를 자동으로 만들어 줍니다.

> 👨‍🍳 **`range(시작, 끝, 건너뛸 크기)` 마스터하기**
> * `range(5)` : `0`부터 `4`까지 **총 5번** 주문서 발행 (`0, 1, 2, 3, 4`)
> * `range(1, 6)` : **1번부터 6번 전(5)까지** 주문서 발행 (`1, 2, 3, 4, 5`)
> * `range(1, 10, 2)` : 1부터 10 전까지 **2칸씩 건너뛰며** 발행 (`1, 3, 5, 7, 9`)

````python
# 예시: 만두피 3장 빚기 (0부터 3 전까지 총 3번)
for i in range(3):
    print(f"{i}번 만두피 완료!")
````

<br/>
<br/>
<br/>


* **2-3 (while 문): 조건이 참(True)인 동안 끝없이 반복하기**
* `while`문은 조건식(`stir <= 3`)이 맞으면 계속 안쪽 코드를 실행합니다.
* **🚨 중요 셰프 팁:** 국물을 저으면서 횟수를 세지 않으면(`stir += 1` 누락), 컴퓨터는 밤새도록 국물만 젓다가 멈추지 않는 **'무한 루프'** 상태에 빠져 주방(프로그램)이 폭발합니다!



````python
stir = 1             # 현재 저은 횟수 (시작점)
while stir <= 3:     # 3번 이하로 저었을 때까지만 반복해라! (종료 조건)
    print("휘젓기!")
    stir += 1        # 중요: 한 번 저었으니 횟수를 1 올린다! (변화 조건)
````

## 🟥 3. 셰프의 비상 수단: `break` / `continue`

> **"손님이 나갔으니 조리 즉시 중단(`break`)! 불량 재료는 건너뛰고 다음 재료 손질(`continue`)!"**

### 🍳 Basic 실습 & 세부 설명

* **3-1 (break): 하던 일을 때려치우고 반복문 탈출!**
* 리스트를 돌며 재료를 넣다가 상한 음식을 발견하는 순간, 뒤에 남은 재료가 무엇이든 상관없이 `for`문 전체를 강제로 부수고 끝냅니다.




````python
basket = ["감자", "상한고기", "양파"]
for food in basket:
    if food == "상한고기":
        print("🚨 멈춰! 상한고기 발견! 조리 전체 중단!")
        break # 이 명령을 만나는 순간, 뒤에 있는 '양파'는 보지도 않고 반복문 탈출!
    print(f"{food} 투하")
````


<br/>
<br/>
<br/>

* **3-2 (continue): 이번 차례만 패스하고 다음 차례로 점프!**
* 특정 재료를 만났을 때, 아래에 있는 코드(`print`)를 실행하지 않고 곧바로 다음 재료를 꺼내러 맨 위줄(`for`)로 새로 시작합니다. (탈출하는 게 아니라 **이번 한 번만 건너뛰는 것**입니다.)





````python
toppings = ["치즈", "피클", "페페로니"]
for top in toppings:
    if top == "피클":
        print("(피클은 빼달라고 하셨지? 건너뛰자!)")
        continue # 아래 코드를 무시하고 다음 토핑인 '페페로니'를 꺼내러 올라감!
    print(f"{top} 올리기")
````

### 🚀 Mission Challenge (실습용)

* **Q3. 자동 짝수 재료 손질하기**
* **설명:** 1번부터 10번 재료 중 `continue`를 활용하여 **홀수 번호는 패스하고, 짝수 번호 재료만** 골라 손질하는 코드를 완성하세요.
* **힌트:** `i % 2 != 0` 식은 숫자를 2로 나눈 나머지가 0이 아닐 때, 즉 **홀수**일 때 참이 됩니다.







````python
# [미션] 빈칸 안에 들어갈 알맞은 제어 키워드를 적어보세요!
for i in range(1, 11): # 1부터 10까지 숫자가 나옵니다.
    if i % 2 != 0:     # 만약 홀수 번호 재료라면?
        continue       # 아래 손질 코드를 실행하지 말고 다음 숫자로 패스! 건너뛰기!

    print(f"재료 번호 {i}번 손질 완료! 🔪")
````
