#### [문제 3] OX 퀴즈: 라이브러리의 속마음 ⭕❌

다음 설명이 맞으면 **O**, 틀리면 **X**를 선택하세요.

1. **Pandas는 NumPy와 완전히 무관하게 독립적으로 처음부터 새롭게 개발된 라이브러리이다.** (**`   `**)
2. **TensorFlow로 딥러닝 모델을 학습시키려면, 데이터를 먼저 모델이 연산하기 적합한 수치적 배열(텐서) 형태로 변환해야 한다.** (**`   `**)




> **정답 순서:**
1. **정답: X**
* **해설:** Pandas는 **NumPy를 기반(Foundation)으로 그 위에 구축**되었습니다. NumPy의 강력한 배열 연산 기능이 밑바탕에 깔려 있기 때문에 Pandas가 대용량 데이터를 빠르게 처리할 수 있는 것입니다.


2. **정답: O**
* **해설:** 딥러닝 엔진인 TensorFlow는 수치적인 연산을 수행하므로, Pandas로 정제된 데이터를 **NumPy 배열 형태의 수치 데이터로 변환**하여 입력해주어야 안전하게 학습이 이루어집니다.
