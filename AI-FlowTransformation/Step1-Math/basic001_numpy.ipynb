#### 1. 셰프의 마법 도구: 함수 만들기

딥러닝은 수많은 함수의 연결입니다. 재료를 넣으면 결과를 뱉어내는 자동 기계를 만들어 봅시다.


````python
# [def] 정식 셰프: 복잡한 로직을 담을 때
def power_up(x):
    return x**2 + 10

# [lambda] 패스트푸드: 한 줄로 끝내는 간편함
fast_power = lambda x: x**2 + 10

print(f"def 사용: {power_up(5)}, lambda 사용: {fast_power(5)}")
````

🦍 딥러닝에서의 역할 (1):
**왜 필요할까요?** **똑같은 계산을 계속 반복해야 할 때, 매번 똑같이 쓰는 건 너무 힘들고 실수하기도 쉬워요. 함수는 계산 과정을 딱 한 번만 정해두고, 필요할 때마다 재료만 바꿔서 바로 쓸 수 있게 해줍니다.**

**딥러닝에서의 역할:** **딥러닝 모델은 수많은 작은 계산(함수)들이 모여서 만들어져요.** 예를 들어, `더하기`, `곱하기` 같은 간단한 계산들을 묶어서 `평균 구하기` 같은 좀 더 큰 계산으로 만드는 거죠. 이렇게 하면 복잡한 모델도 한 부분씩 나눠서 만들고 고칠 수 있어요.

🦍 더 쉽게 풀어보는 딥러닝 예시 (2):
친구들 열 명의 시험 점수를 모두 2배로 올려주는 계산을 한다고 해봐요. `점수 * 2`를 열 번 쓰는 대신, `점수를 2배로 만드는 함수`를 하나 만들어두고, 친구 점수만 넣어주면 알아서 2배가 되는 식이에요. **딥러닝도 이런 식으로 수많은 데이터를 한 번에 처리할 때 함수를 써요.**

🦍 더더 쉽게 풀어보는 딥러닝 예시 (3):
엄마가 매일 아침 아이 도시락을 싸신다고 가정해봐요. 매번 김밥을 처음부터 만드는 대신, `밥 준비하기`, `재료 썰기`, `김밥 말기`처럼 각각의 단계를 함수로 만들어 놓으면, **필요할 때마다 그 함수를 불러서 더 빠르고 정확하게 도시락을 만들 수 있죠.**

#### 3. 재료 번호표 뽑기 (range)

재료가 많을 땐 `range`로 번호표를! **주의: 마지막 숫자는 절대 포함 안 해요!**

````python
# range(숫자1) - 시작 0, 끝 5 직전까지
print(list(range(5)))        # [0, 1, 2, 3, 4]
# range(숫자1, 숫자2) - 시작 1, 끝 5 직전까지
print(list(range(1, 5)))     # [1, 2, 3, 4]
# range(숫자1, 숫자2, 숫자3) - 시작 1, 끝 5 직전까지, 스텝 2
print(list(range(1, 5, 2)))  # [1, 3]
print(list(range(1, 6, 2)))  # [1, 3, 5]
````

🦍 딥러닝에서의 역할 (1):
**왜 필요할까요?** **여러 번 반복해서 뭔가를 해야 할 때, '첫 번째부터 다섯 번째까지'처럼 순서대로 정해진 횟수만큼 동작을 시키고 싶을 때가 많아요. `range`는 이런 반복 작업을 쉽게 할 수 있게 해주는 '번호표 발급기' 같은 역할을 합니다.**

**딥러닝에서의 역할:** **딥러닝 모델은 데이터를 여러 번 반복해서 학습하는데, 이때 `range`를 사용해서 '몇 번 반복할지'를 정해요.** 또는 데이터의 특정 부분만 가져와서 처리할 때도 `range`로 순서를 지정하죠.

🦍 더 쉽게 풀어보는 딥러닝 예시 (2):
덧셈 문제 10개를 풀어야 한다면, '1번 문제, 2번 문제, ..., 10번 문제' 이렇게 순서대로 푸는 것과 같아요. **딥러닝 모델이 100장의 사진을 학습할 때, `range(100)`을 써서 0번째 사진부터 99번째 사진까지 순서대로 가져와서 학습하라고 명령할 수 있어요.**

🦍 더더 쉽게 풀어보는 딥러닝 예시 (3):
놀이공원에서 놀이기구를 탈 때 '1번 칸부터 5번 칸까지 탑승'이라고 안내하는 것과 같아요. **`range(1, 6)`을 하면 1, 2, 3, 4, 5 이렇게 순서대로 칸을 지정해 주는 거죠.**

#### 4. 모양 맞추기: Numpy Reshape

똑같은 24개의 재료라도 어떤 접시(Shape)에 담느냐에 따라 요리 모양이 달라집니다.

````python
A = np.array(range(24))
print("기본값" , A)
print("shape" , A.shape)
print("\n\n")

B = A.reshape(2, 12) # 2단 12열
C = A.reshape(3, 8)  # 3단 8열
D = A.reshape(4, -1) # 4단, 나머지는 알아서 채워줘!
print("24개 재료의 변신:\n", B, "\n\n", C, "\n\n", D)
````

🦍 딥러닝에서의 역할 (1):
**왜 필요할까요?** **같은 숫자라도 어떻게 줄을 세우느냐에 따라 우리가 이해하는 방식이나 계산하는 방식이 달라져요. 예를 들어, 1부터 24까지의 숫자를 한 줄로 늘어놓는 것보다 4줄 6칸으로 나누어 놓으면, 마치 구구단표처럼 어떤 관계를 더 쉽게 찾거나 특정 규칙을 적용하기 편해지죠.**

**딥러닝에서의 역할:** **딥러닝 모델은 입력되는 데이터의 '모양'에 매우 민감해요.** 특히 이미지를 처리하는 모델은 데이터를 '가로 세로 픽셀 수'와 '색상 정보'가 있는 특정한 2차원 또는 3차원 형태로 받아야 해요. `reshape`는 어떤 형태로든 들어온 데이터를 모델이 이해할 수 있는 모양으로 바꿔주는 중요한 작업입니다.

🦍 더 쉽게 풀어보는 딥러닝 예시 (2):
사진을 컴퓨터에 저장할 때는 모든 픽셀 정보가 길게 한 줄로 저장될 수 있어요. 하지만 우리가 그 사진을 볼 때는 가로, 세로가 있는 그림으로 보잖아요? **`reshape`는 길게 늘어진 픽셀 정보를 가로 세로에 맞춰서 다시 배치하는 것과 같아요. 그래야 딥러닝 모델이 '이건 사람 얼굴이네!' 하고 알아볼 수 있죠.**

🦍 더더 쉽게 풀어보는 딥러닝 예시 (3):
24명의 학생이 운동장에 줄지어 서 있어요. 이 학생들을 1열로 길게 세울 수도 있고, 4명씩 6줄로 나누어 세울 수도 있죠. **`reshape`는 학생들을 다시 줄 세우는 것처럼 데이터의 '배열 방식'을 바꾸는 거라고 생각하면 돼요.**

#### 5. 정밀 계량: arange와 linspace

`arange`는 **간격(step)** 중시, `linspace`는 **개수(num)** 중시! 맛의 정밀도를 조절하는 기법입니다.


````python
# arange: 1부터 10 전까지 2씩 띄어서
A1 = np.arange(1, 10, 2)
# linspace: 1부터 9까지를 똑같은 5등분으로
A2 = np.linspace(1, 9, 5)

A3 = np.linspace(1,10,5)
print('간격 중심(arange):', A1, '\n\n개수 중심(linspace):', A2  ,'\n\n개수 중심(linspace):', A3)
````

🦍 딥러닝에서의 역할 (1):
**왜 필요할까요?** **어떤 때는 일정한 간격으로 숫자를 만들고 싶고(예: 2, 4, 6, 8...), 어떤 때는 특정 범위 안에서 원하는 개수만큼 숫자를 똑같이 나눠서 만들고 싶을 때(예: 0부터 10까지 5개 숫자)가 있어요. 이 두 가지 방식은 정확한 계산이나 분석을 위해 꼭 필요하답니다.**

**딥러닝에서의 역할:** **딥러닝 모델에서 함수의 변화를 시각화하거나, 특정 범위 내에서 테스트할 입력 값을 생성할 때 이 함수들이 유용하게 쓰여요.** 예를 들어, 활성화 함수가 어떤 입력 값에 대해 어떻게 반응하는지 연속적으로 확인하고 싶을 때 사용하죠.

🦍 더 쉽게 풀어보는 딥러닝 예시 (2):
시험 점수를 0점부터 100점까지 10점 단위로 만들려면 `arange`를 쓸 수 있어요 (0, 10, 20, ..., 100). **반면, 0점부터 100점까지 딱 5개의 점수로 똑같이 나누어 대표값을 만들고 싶으면 `linspace`를 쓸 수 있죠 (0, 25, 50, 75, 100).**

🦍 더더 쉽게 풀어보는 딥러닝 예시 (3):
빵을 만들 때, `arange`는 '밀가루 2컵, 설탕 2스푼, 소금 2꼬집'처럼 정확히 2단위로 계량하는 것과 비슷해요. **`linspace`는 케이크 하나를 정확히 6등분해서 나눠 먹고 싶을 때, 각 조각의 크기를 똑같이 만드는 것과 같죠.**

* **실제 생성 값:** `[0., 2., 4., 6., 8., 10.]`

##### [문제 6] 응용: 셰프의 계산기 🧮

앞서 배운 정밀 계량 공식을 활용해 보는 문제입니다.
`np.linspace(0, 20, 5)`를 사용하면 어떤 숫자 5개가 생성될지, 아래 공식을 이용해 간격을 구하고 최종 배열을 직접 계산해 보세요.

$$\text{간격} = \frac{\text{끝값} - \text{시작값}}{\text{개수} - 1}$$



* **간격 계산:**

$$\text{간격} = \frac{20 - 0}{5 - 1} = \frac{20}{4} = 5.0$$


* **최종 결과:** 시작값 0부터 5.0씩 더해가며 5개의 숫자를 채웁니다.

$$\rightarrow \text{[0.0, 5.0, 10.0, 15.0, 20.0]}$$

##### 1. '구간'과 '숫자'의 차이

`np.linspace(1, 10, 5)`는 "1부터 10까지 5개의 숫자를 찍어라"는 뜻입니다.

* **숫자 5개를 찍으려면:** 1(첫 번째), ?, ?, ?, 10(다섯 번째)
* **사이사이의 틈(구간)은 몇 개가 생기나요?**
* 1번과 2번 사이(1구간)
* 2번과 3번 사이(2구간)
* 3번과 4번 사이(3구간)
* 4번과 5번 사이(4구간)
* **결과적으로 4개의 구간**이 생깁니다.



---

##### 2. 왜 (5 - 1)로 나누나요?

전체 거리(10 - 1 = 9)를 똑같은 크기의 '막대기' 4개로 쪼개야 하니까, **전체 길이 9를 4로 나누는 것**입니다.

* **수식:** $\text{간격} = \frac{\text{끝값} - \text{시작값}}{\text{개수} - 1}$
* **대입:** $\frac{10 - 1}{5 - 1} = \frac{9}{4} = 2.25$

그래서 숫자가 1에서 시작해서 2.25씩 커지는 것입니다.

* 1번째 숫자: **1.0**
* 2번째 숫자: 1.0 + 2.25 = **3.25**
* 3번째 숫자: 3.25 + 2.25 = **5.5**
* 4번째 숫자: 5.5 + 2.25 = **7.75**
* 5번째 숫자: 7.75 + 2.25 = **10.0**

#### 6. 랜덤 시즈닝과 맛 평가 (random & plot)

때로는 무작위(`random`) 시즈닝이 창의적인 요리를 만듭니다. 균형 잡힌 맛을 낼 때는 정규분포(`normal`) 시즈닝을 사용하죠.


🦍 딥러닝에서의 역할 (1):
**왜 필요할까요?** **우리가 모르는 새로운 상황을 예측하거나, 데이터 속에 숨어있는 규칙을 찾을 때, '아무렇게나' 숫자를 뽑아보거나(랜덤) 그 숫자들을 그림으로 그려서(플롯) 살펴보는 것이 큰 도움이 돼요. 눈으로 봐야 무엇이 문제인지, 어떤 패턴이 있는지 알 수 있으니까요.**

**딥러닝에서의 역할:** **딥러닝 모델이 처음 학습을 시작할 때, 가중치(모델의 핵심 '조절값')는 무작위로 설정돼요. 이는 모델이 다양한 가능성을 탐색하게 돕죠.** 그리고 **모델이 학습한 결과를 눈으로 확인하려면 그래프(플롯)가 필수적이에요.**

🦍 더 쉽게 풀어보는 딥러닝 예시 (2):
동전 던지기를 100번 했다고 해봐요. 앞면과 뒷면이 나올 횟수가 대충 반반이라는 걸 알지만, 정확히 몇 번씩 나올지는 무작위죠. **`random`은 이런 동전 던지기 결과를 만들어내는 것과 같아요.** 그리고 이 결과를 막대그래프나 선 그래프로 그리면(플롯), **'아, 앞면이 조금 더 많이 나왔네' 하고 한눈에 볼 수 있죠.**

🦍 더더 쉽게 풀어보는 딥러닝 예시 (3):
새로운 맛집의 손님 만족도를 조사한다고 해봐요. 무작위로 100명의 손님을 뽑아서(랜덤) 별점을 주게 해요. 그 별점들을 모아서 그래프로 그리면(플롯), **'대부분의 손님은 별 4개를 줬지만, 별 1개를 준 사람도 몇 명 있네?' 하고 전체적인 만족도를 파악할 수 있어요.**

### 💡 확인 문제: 6. 랜덤 시즈닝과 맛 평가 (random & plot)

Q1. 딥러닝 모델이 처음 학습을 시작할 때, 가중치는 종종 [ ]으로 설정되어 다양한 가능성을 탐색하게 돕습니다.

Q2. 데이터의 분포나 숨겨진 패턴을 눈으로 확인하려면 `plot`을 사용하여 시각화하는 것이 중요하다. (O/X)

Q3. 동전 던지기 결과를 `random`으로 100번 만들었다고 할 때, 이 결과를 그림으로 그려서 앞면과 뒷면의 횟수를 한눈에 확인하고 싶다면 어떤 도구를 사용해야 할까요?


✅ 정답

Q1. 무작위 (랜덤)

Q2. O

Q3. `matplotlib`의 `plt.plot`이나 `plt.scatter`와 같은 시각화 도구 (또는 단순히 '플롯' 또는 '그래프 그리기'도 정답입니다.)

````python
# 1) 균등분포 랜덤 데이터 생성 및 시각화
A1 = np.random.random(100)
A2 = np.random.rand(100)
A1.sort()
A2.sort()

plt.figure(figsize=(6, 4))
plt.plot(A1, c='b', label='random')
plt.plot(A2, c='r', label='rand')
plt.legend()
plt.title("Random Data Distribution")
plt.show()

# 2) 정규분포 데이터 생성 (평균 -0.5, 표준편차 0.3, 250개)
data_norm = np.random.normal(-0.5, 0.3, 250)
data_norm.sort()

# 결과 시각화 (산점도와 선 그래프)
plt.figure(figsize=(6, 4))
plt.scatter(data_norm, data_norm, c='b', marker='.', label='Scatter')
plt.title("Normal Distribution Scatter")
plt.legend()
plt.show()

plt.figure(figsize=(6, 4))
plt.plot(data_norm, c='b', label='Line')
plt.title("Normal Distribution Line")
plt.legend()
plt.show()
````


#### 1. 왜 중요한가요? (데이터의 '성질'을 파악하는 눈)

현실 세계의 데이터는 무작위로 흩어져 있습니다. 우리가 가진 코드는 "무작위 데이터 속에 숨겨진 규칙(패턴)을 찾아내는 연습"을 하는 것입니다.

* **무작위 데이터(`random`)**: 0과 1 사이에서 아무 숫자나 뽑아보면, 처음엔 뒤죽박죽입니다. 하지만 `sort()`로 정렬하고 그래프를 그리면, "아, 데이터가 이렇게 골고루 퍼져 있구나"라는 **분포의 특징**이 한눈에 들어옵니다.
* **정규분포(`normal`)**: 자연계의 거의 모든 데이터(키, 몸무게, 시험 점수 등)는 정규분포(종 모양 곡선)를 따릅니다. 이 분포를 이해하면, 미래에 어떤 데이터가 들어왔을 때 "이건 평범한 데이터다" 혹은 "이건 아주 특이한 데이터(이상치)다"를 판단하는 **통계적 기준**을 세울 수 있습니다.

#### 2. 왜 쉬운가요? (시각화의 마법)

데이터를 눈으로 보지 않으면 그냥 '숫자 덩어리'일 뿐입니다. 하지만 `matplotlib`(`plt`)을 쓰면 복잡한 통계 계산을 **단 3줄의 코드로 그림**으로 바꿀 수 있습니다.

* `plt.plot`: 데이터가 전체적으로 어떻게 흐르는지(추세) 보여줍니다.
* `plt.scatter`: 데이터 하나하나가 어디에 찍히는지(밀도) 보여줍니다.
* **재미 포인트:** 그냥 숫자만 출력하면 지루하지만, `plt.show()`를 누르는 순간 **내가 만든 데이터가 예술적인 그래프로 변하는 쾌감**이 있습니다. 마치 데이터에 생명을 불어넣는 느낌이죠!
