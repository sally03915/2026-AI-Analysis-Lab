#### 1. 셰프의 마법 도구: 함수 만들기

딥러닝은 수많은 함수의 연결입니다. 재료를 넣으면 결과를 뱉어내는 자동 기계를 만들어 봅시다.



*   **함수 다이어그램:** `입력 -> [함수 상자] -> 출력`
*   **요리 기계 비유:** 재료(x)를 넣으면 자동으로 요리(x**2 + 10)가 나와요!

001-1_1.png

```python
# [def] 정식 셰프: 복잡한 로직을 담을 때
def power_up(x):
    return x**2 + 10

# [lambda] 패스트푸드: 한 줄로 끝내는 간편함
fast_power = lambda x: x**2 + 10

print(f"def 사용: {power_up(5)}, lambda 사용: {fast_power(5)}")
```

🦍 딥러닝에서의 역할 (1):
**왜 필요할까요?** **똑같은 계산을 계속 반복해야 할 때, 매번 똑같이 쓰는 건 너무 힘들고 실수하기도 쉬워요. 함수는 계산 과정을 딱 한 번만 정해두고, 필요할 때마다 재료만 바꿔서 바로 쓸 수 있게 해줍니다.**

**딥러닝에서의 역할:** **딥러닝 모델은 수많은 작은 계산(함수)들이 모여서 만들어져요.** 예를 들어, `더하기`, `곱하기` 같은 간단한 계산들을 묶어서 `평균 구하기` 같은 좀 더 큰 계산으로 만드는 거죠. 이렇게 하면 복잡한 모델도 한 부분씩 나눠서 만들고 고칠 수 있어요.

🦍 더 쉽게 풀어보는 딥러닝 예시 (2):
친구들 열 명의 시험 점수를 모두 2배로 올려주는 계산을 한다고 해봐요. `점수 * 2`를 열 번 쓰는 대신, `점수를 2배로 만드는 함수`를 하나 만들어두고, 친구 점수만 넣어주면 알아서 2배가 되는 식이에요. **딥러닝도 이런 식으로 수많은 데이터를 한 번에 처리할 때 함수를 써요.**

🦍 더더 쉽게 풀어보는 딥러닝 예시 (3):
엄마가 매일 아침 아이 도시락을 싸신다고 가정해봐요. 매번 김밥을 처음부터 만드는 대신, `밥 준비하기`, `재료 썰기`, `김밥 말기`처럼 각각의 단계를 함수로 만들어 놓으면, **필요할 때마다 그 함수를 불러서 더 빠르고 정확하게 도시락을 만들 수 있죠.**

##### [문제 1] 함수 만들기: def vs lambda 🪄

입력값 $x$에 **3을 곱하고 5를 더하는** 자동 기계(함수)를 만들려고 합니다.

1. `def`를 사용하여 `triple_add`라는 이름의 함수를 만드세요.
2. `lambda`를 사용하여 같은 기능을 하는 `fast_triple`이라는 변수를 만드세요.
3. 두 함수에 각각 `10`을 넣었을 때의 결과값을 출력하세요.


<br/>
<br/>
<br/>

```python
# 1. def 정석 함수
def triple_add(x):
    return x * 3 + 5

# 2. lambda 간편 함수
fast_triple = lambda x: x * 3 + 5

# 3. 결과 출력
print("def 결과:", triple_add(10))     # 결과: 35
print("lambda 결과:", fast_triple(10)) # 결과: 35
```

#### 2. 재료 손질하기 (데이터 생성)

컴퓨터가 알아듣게 정수(`int`), 실수(`float`), 목록(`list`)으로 재료를 다듬습니다.


*   **재료 매칭:** `int`에는 사과 3개 , `float`에는 자(ruler) 또는 온도계 , `list`에는 장바구니에 여러 과일이 담긴 개념
*   **재료의 이름표:** 재료마다 '개수', '무게', '종류' 등 정확한 이름표를 붙여야 요리사가 헷갈리지 않는 것처럼, 데이터에도 정확한 타입을 부여해야 컴퓨터가 올바르게 처리합니다.

001-2.png




```python
# 정수(int): 77을 세 가지 방식으로 생성
A1 = 77; A2 = int('77'); A3 = int()
print('A1:', A1, 'A2:', A2, 'A3:', A3)

# 실수(float): 3.4를 세 가지 방식으로 생성
B1 = 3.4; B2 = float('3.4'); B3 = float()
print('B1:', B1, 'B2:', B2, 'B3:', B3)

# 목록(list): 여러 재료를 바구니에
C1 = [0,1,2,3];   C2 = list(range(4)); C3 = list()
print('C1:', C1, 'C2:', C2, 'C3:', C3)
```

**딥러닝에서의 역할:** **딥러닝 모델은 여러 종류의 데이터를 이해하고 처리해야 해요.** 어떤 데이터는 '개수'처럼 딱 떨어지는 정수(int)로, 어떤 데이터는 '길이'나 '온도'처럼 소수점이 있는 실수(float)로 표현해야 하죠. 여러 개의 정보를 한 번에 묶어 전달할 때는 목록(list)을 사용해요.

#### 3. 재료 번호표 뽑기 (range)

재료가 많을 땐 `range`로 번호표를! **주의: 마지막 숫자는 절대 포함 안 해요!**


*   **번호표 발급기:** `range` 함수가 0부터 시작하여 순서대로 번호표를 톡톡 발급
*   **컨베이어 벨트:** 공장에서 컨베이어 벨트 위에 놓인 제품들이 '1번 제품, 2번 제품...' 순서대로 다음 공정으로 넘어가는  반복 작업의 순차성

001-3.png


```python
# range(숫자1) - 시작 0, 끝 5 직전까지
print(list(range(5)))        # [0, 1, 2, 3, 4]
# range(숫자1, 숫자2) - 시작 1, 끝 5 직전까지
print(list(range(1, 5)))     # [1, 2, 3, 4]
# range(숫자1, 숫자2, 숫자3) - 시작 1, 끝 5 직전까지, 스텝 2
print(list(range(1, 5, 2)))  # [1, 3]
print(list(range(1, 6, 2)))  # [1, 3, 5]
```

🦍 딥러닝에서의 역할 (1):
**왜 필요할까요?** **여러 번 반복해서 뭔가를 해야 할 때, '첫 번째부터 다섯 번째까지'처럼 순서대로 정해진 횟수만큼 동작을 시키고 싶을 때가 많아요. `range`는 이런 반복 작업을 쉽게 할 수 있게 해주는 '번호표 발급기' 같은 역할을 합니다.**

**딥러닝에서의 역할:** **딥러닝 모델은 데이터를 여러 번 반복해서 학습하는데, 이때 `range`를 사용해서 '몇 번 반복할지'를 정해요.** 또는 데이터의 특정 부분만 가져와서 처리할 때도 `range`로 순서를 지정하죠.

🦍 더 쉽게 풀어보는 딥러닝 예시 (2):
덧셈 문제 10개를 풀어야 한다면, '1번 문제, 2번 문제, ..., 10번 문제' 이렇게 순서대로 푸는 것과 같아요. **딥러닝 모델이 100장의 사진을 학습할 때, `range(100)`을 써서 0번째 사진부터 99번째 사진까지 순서대로 가져와서 학습하라고 명령할 수 있어요.**

🦍 더더 쉽게 풀어보는 딥러닝 예시 (3):
놀이공원에서 놀이기구를 탈 때 '1번 칸부터 5번 칸까지 탑승'이라고 안내하는 것과 같아요. **`range(1, 6)`을 하면 1, 2, 3, 4, 5 이렇게 순서대로 칸을 지정해 주는 거죠.**

#### 4. 모양 맞추기: Numpy Reshape

똑같은 24개의 재료라도 어떤 접시(Shape)에 담느냐에 따라 요리 모양이 달라집니다.

*   **블록 쌓기/재배치:** 레고 블록들이 처음에는 일렬로 놓여있다가, `reshape`를 통해 2x12, 3x8, 4x6 등의 형태로 재배치
*   **좌석 배치도:** 극장 좌석이 처음에는 복도에 일렬로 대기하다가, '3줄 4칸' 자리로 안내받아 앉는 물리적 재배치

001-4.png

```python
A = np.array(range(24))
print("기본값" , A)
print("shape" , A.shape)
print("\n\n")

B = A.reshape(2, 12) # 2단 12열
C = A.reshape(3, 8)  # 3단 8열
D = A.reshape(4, -1) # 4단, 나머지는 알아서 채워줘!
print("24개 재료의 변신:\n", B, "\n\n", C, "\n\n", D)
```

🦍 딥러닝에서의 역할 (1):
**왜 필요할까요?** **같은 숫자라도 어떻게 줄을 세우느냐에 따라 우리가 이해하는 방식이나 계산하는 방식이 달라져요. 예를 들어, 1부터 24까지의 숫자를 한 줄로 늘어놓는 것보다 4줄 6칸으로 나누어 놓으면, 마치 구구단표처럼 어떤 관계를 더 쉽게 찾거나 특정 규칙을 적용하기 편해지죠.**

**딥러닝에서의 역할:** **딥러닝 모델은 입력되는 데이터의 '모양'에 매우 민감해요.** 특히 이미지를 처리하는 모델은 데이터를 '가로 세로 픽셀 수'와 '색상 정보'가 있는 특정한 2차원 또는 3차원 형태로 받아야 해요. `reshape`는 어떤 형태로든 들어온 데이터를 모델이 이해할 수 있는 모양으로 바꿔주는 중요한 작업입니다.

#### 5. 정밀 계량: arange와 linspace

`arange`는 **간격(step)** 중시, `linspace`는 **개수(num)** 중시! 맛의 정밀도를 조절하는 기법입니다.


*   **눈금자와 분할:**
    *   `arange`: 일정한 간격(예: 2cm마다)으로 눈금이 표시된 자(ruler)  '간격 중심'
    *   `linspace`: 0cm부터 10cm까지의 선을 '정확히 5등분'하는  '개수 중심'
*   **빵 자르기:**
    *   `arange`: 빵을 '2cm 두께로' 일정하게 자르는 것
    *   `linspace`: 빵 하나를 '정확히 5조각으로' 나누는 것


    

001-5.png

```python

# arange: 1부터 10 전까지 2씩 띄어서
A1 = np.arange(1, 10, 2)
# linspace: 1부터 9까지를 똑같은 5등분으로
A2 = np.linspace(1, 9, 5)

A3 = np.linspace(1,10,5)
print('간격 중심(arange):', A1, '\n\n개수 중심(linspace):', A2  ,'\n\n개수 중심(linspace):', A3)
```

🦍 딥러닝에서의 역할 (1):
**왜 필요할까요?** **어떤 때는 일정한 간격으로 숫자를 만들고 싶고(예: 2, 4, 6, 8...), 어떤 때는 특정 범위 안에서 원하는 개수만큼 숫자를 똑같이 나눠서 만들고 싶을 때(예: 0부터 10까지 5개 숫자)가 있어요. 이 두 가지 방식은 정확한 계산이나 분석을 위해 꼭 필요하답니다.**

**딥러닝에서의 역할:** **딥러닝 모델에서 함수의 변화를 시각화하거나, 특정 범위 내에서 테스트할 입력 값을 생성할 때 이 함수들이 유용하게 쓰여요.** 예를 들어, 활성화 함수가 어떤 입력 값에 대해 어떻게 반응하는지 연속적으로 확인하고 싶을 때 사용하죠.

🦍 더 쉽게 풀어보는 딥러닝 예시 (2):

🦍 더더 쉽게 풀어보는 딥러닝 예시 (3):

#####  [문제 6] 응용: 셰프의 계산기 🧮

앞서 배운 정밀 계량 공식을 활용해 보는 문제입니다.
`np.linspace(0, 20, 5)`를 사용하면 어떤 숫자 5개가 생성될지, 아래 공식을 이용해 간격을 구하고 최종 배열을 직접 계산해 보세요.

$$\text{간격} = \frac{\text{끝값} - \text{시작값}}{\text{개수} - 1}$$



* **간격 계산:**

$$\text{간격} = \frac{20 - 0}{5 - 1} = \frac{20}{4} = 5.0$$


* **최종 결과:** 시작값 0부터 5.0씩 더해가며 5개의 숫자를 채웁니다.

$$\rightarrow \text{[0.0, 5.0, 10.0, 15.0, 20.0]}$$

##### 1. '구간'과 '숫자'의 차이

`np.linspace(1, 10, 5)`는 "1부터 10까지 5개의 숫자를 찍어라"는 뜻입니다.

* **숫자 5개를 찍으려면:** 1(첫 번째), ?, ?, ?, 10(다섯 번째)
* **사이사이의 틈(구간)은 몇 개가 생기나요?**
* 1번과 2번 사이(1구간)
* 2번과 3번 사이(2구간)
* 3번과 4번 사이(3구간)
* 4번과 5번 사이(4구간)
* **결과적으로 4개의 구간**이 생깁니다.



---

##### 2. 왜 (5 - 1)로 나누나요?

전체 거리(10 - 1 = 9)를 똑같은 크기의 '막대기' 4개로 쪼개야 하니까, **전체 길이 9를 4로 나누는 것**입니다.

* **수식:** $\text{간격} = \frac{\text{끝값} - \text{시작값}}{\text{개수} - 1}$
* **대입:** $\frac{10 - 1}{5 - 1} = \frac{9}{4} = 2.25$

그래서 숫자가 1에서 시작해서 2.25씩 커지는 것입니다.

* 1번째 숫자: **1.0**
* 2번째 숫자: 1.0 + 2.25 = **3.25**
* 3번째 숫자: 3.25 + 2.25 = **5.5**
* 4번째 숫자: 5.5 + 2.25 = **7.75**
* 5번째 숫자: 7.75 + 2.25 = **10.0**

#### 6. 랜덤 시즈닝과 맛 평가 (random & plot)

때로는 무작위(`random`) 시즈닝이 창의적인 요리를 만듭니다. 균형 잡힌 맛을 낼 때는 정규분포(`normal`) 시즈닝을 사용하죠.



*   **주사위 던지기 그림:** `random` 함수가 주사위를 던져 무작위 숫자를 생성
*   **데이터 엑스레이/지도:** `plot`이 데이터의 보이지 않는 속성을 '엑스레이'처럼 보여주기, 복잡한 지도를 '간단한 그림'으로 압축보여주기

001-6.png

🦍 딥러닝에서의 역할 (1):
**왜 필요할까요?** **우리가 모르는 새로운 상황을 예측하거나, 데이터 속에 숨어있는 규칙을 찾을 때, '아무렇게나' 숫자를 뽑아보거나(랜덤) 그 숫자들을 그림으로 그려서(플롯) 살펴보는 것이 큰 도움이 돼요. 눈으로 봐야 무엇이 문제인지, 어떤 패턴이 있는지 알 수 있으니까요.**

**딥러닝에서의 역할:** **딥러닝 모델이 처음 학습을 시작할 때, 가중치(모델의 핵심 '조절값')는 무작위로 설정돼요. 이는 모델이 다양한 가능성을 탐색하게 돕죠.** 그리고 **모델이 학습한 결과를 눈으로 확인하려면 그래프(플롯)가 필수적이에요.**

🦍 더 쉽게 풀어보는 딥러닝 예시 (2):
동전 던지기를 100번 했다고 해봐요. 앞면과 뒷면이 나올 횟수가 대충 반반이라는 걸 알지만, 정확히 몇 번씩 나올지는 무작위죠. **`random`은 이런 동전 던지기 결과를 만들어내는 것과 같아요.** 그리고 이 결과를 막대그래프나 선 그래프로 그리면(플롯), **'아, 앞면이 조금 더 많이 나왔네' 하고 한눈에 볼 수 있죠.**

🦍 더더 쉽게 풀어보는 딥러닝 예시 (3):
새로운 맛집의 손님 만족도를 조사한다고 해봐요. 무작위로 100명의 손님을 뽑아서(랜덤) 별점을 주게 해요. 그 별점들을 모아서 그래프로 그리면(플롯), **'대부분의 손님은 별 4개를 줬지만, 별 1개를 준 사람도 몇 명 있네?' 하고 전체적인 만족도를 파악할 수 있어요.**

```python
# 1) 균등분포 랜덤 데이터 생성 및 시각화
A1 = np.random.random(100)
A2 = np.random.rand(100)
A1.sort()
A2.sort()

plt.figure(figsize=(6, 4))
plt.plot(A1, c='b', label='random')
plt.plot(A2, c='r', label='rand')
plt.legend()
plt.title("Random Data Distribution")
plt.show()

# 2) 정규분포 데이터 생성 (평균 -0.5, 표준편차 0.3, 250개)
data_norm = np.random.normal(-0.5, 0.3, 250)
data_norm.sort()

# 결과 시각화 (산점도와 선 그래프)
plt.figure(figsize=(6, 4))
plt.scatter(data_norm, data_norm, c='b', marker='.', label='Scatter')
plt.title("Normal Distribution Scatter")
plt.legend()
plt.show()

plt.figure(figsize=(6, 4))
plt.plot(data_norm, c='b', label='Line')
plt.title("Normal Distribution Line")
plt.legend()
plt.show()
```


#### 1. 왜 중요한가요? (데이터의 '성질'을 파악하는 눈)

현실 세계의 데이터는 무작위로 흩어져 있습니다. 우리가 가진 코드는 "무작위 데이터 속에 숨겨진 규칙(패턴)을 찾아내는 연습"을 하는 것입니다.

* **무작위 데이터(`random`)**: 0과 1 사이에서 아무 숫자나 뽑아보면, 처음엔 뒤죽박죽입니다. 하지만 `sort()`로 정렬하고 그래프를 그리면, "아, 데이터가 이렇게 골고루 퍼져 있구나"라는 **분포의 특징**이 한눈에 들어옵니다.
* **정규분포(`normal`)**: 자연계의 거의 모든 데이터(키, 몸무게, 시험 점수 등)는 정규분포(종 모양 곡선)를 따릅니다. 이 분포를 이해하면, 미래에 어떤 데이터가 들어왔을 때 "이건 평범한 데이터다" 혹은 "이건 아주 특이한 데이터(이상치)다"를 판단하는 **통계적 기준**을 세울 수 있습니다.

#### 2. 왜 쉬운가요? (시각화의 마법)

데이터를 눈으로 보지 않으면 그냥 '숫자 덩어리'일 뿐입니다. 하지만 `matplotlib`(`plt`)을 쓰면 복잡한 통계 계산을 **단 3줄의 코드로 그림**으로 바꿀 수 있습니다.

* `plt.plot`: 데이터가 전체적으로 어떻게 흐르는지(추세) 보여줍니다.
* `plt.scatter`: 데이터 하나하나가 어디에 찍히는지(밀도) 보여줍니다.
* **재미 포인트:** 그냥 숫자만 출력하면 지루하지만, `plt.show()`를 누르는 순간 **내가 만든 데이터가 예술적인 그래프로 변하는 쾌감**이 있습니다. 마치 데이터에 생명을 불어넣는 느낌이죠!

#### 7. 상황에 따라 요리법 바꾸기: 제어문 (Control Flow)

'만약 이렇게 되면 저렇게 하고, 아니면 다른 방법으로 해' 혹은 '이걸 정해진 횟수만큼 반복해'처럼, 코드의 흐름을 조절하는 기법입니다.

#### ✨ 시각적 피드백 및 직관적 비유 추가 제안:

*   **간단한 흐름도(Flowchart):** `if/elif/else`는 분기점(다이아몬드), `for`와 `while`은 반복 화살표가 있는 간단한 흐름도를 추가하여 코드의 실행 순서를 시각적으로 보여줄 수 있습니다.
*   **주방장의 의사결정/반복 작업 비유:**
    *   `if`: 재료 상태를 보고 "이 재료가 좋으면 A 요리, 아니면 B 요리"를 결정하는 주방장의 모습을 연출합니다.
    *   `for`: "오늘 예약 손님 10명에게 모두 같은 코스 요리를 제공"하는 반복적인 작업을 그림으로 나타냅니다.
    *   `while`: "반죽이 부드러워질 때까지 계속 치대라"는 지시를 따르는 요리사의 모습을 그려 조건이 충족될 때까지 반복하는 과정을 보여줍니다.

```python
# if, elif, else: 재료 상태에 따른 요리법 결정
temperature = 25

if temperature > 30:
    print("날씨가 너무 더워요! 시원한 냉면을 만듭니다.")
elif temperature > 20:
    print("날씨가 좋네요! 파스타를 만듭니다.")
else:
    print("날씨가 쌀쌀하네요. 따뜻한 국물을 끓입니다.")

print("\n")

# for: 정해진 횟수만큼 반복하기 (손님 수에 맞춰 접시 준비)
for i in range(3):
    print(f"손님 {i+1}을 위한 접시를 준비합니다.")

print("\n")

# while: 조건이 만족할 때까지 반복하기 (물이 끓을 때까지 기다리기)
water_temp = 80
while water_temp < 100:
    print(f"물이 {water_temp}도입니다. 끓는 중...")
    water_temp += 5 # 5도씩 온도 상승
print(f"물이 {water_temp}도로 끓었습니다! 요리 시작!")
```

🦍 딥러닝에서의 역할 (1):
**왜 필요할까요?** **요리를 할 때, 재료의 상태(신선도, 양 등)에 따라 조리법을 바꾸거나, 손님 수에 맞춰 반복적으로 음식을 준비해야 할 때가 많죠? 코딩에서도 특정 조건에 따라 다른 동작을 수행하거나, 특정 작업을 여러 번 반복해야 할 때 '흐름 제어'가 필요해요.**

**딥러닝에서의 역할:** **딥러닝 모델의 학습 과정은 제어문의 연속입니다.** 예를 들어, '학습 오차가 특정 값보다 작아지면 학습을 멈춰라' (`if`), '모든 데이터를 100번 반복해서 학습해라' (`for` 또는 Epoch), '오차가 더 이상 줄어들지 않을 때까지 학습을 계속해라' (`while`)와 같이 사용됩니다. 이를 통해 모델의 효율적인 학습과 최적화를 이끌어낼 수 있습니다.

🦍 더 쉽게 풀어보는 딥러닝 예시 (2):
로봇 청소기가 집을 청소한다고 생각해 보세요. '방에 먼지가 있으면 청소하고(if), 모든 방을 다 돌 때까지(for) 계속 청소해라'와 같이 동작하죠. **딥러닝 모델도 데이터를 처리하고 학습할 때, 이처럼 '조건에 따라', '반복적으로' 작업을 수행합니다.**

🦍 더더 쉽게 풀어보는 딥러닝 예시 (3):
레스토랑 주방에서 주문을 처리하는 것과 같아요. '손님이 스테이크를 주문했으면 스테이크를 굽고, 파스타를 주문했으면 파스타를 만들어라(if/elif/else)'. 그리고 '오늘 들어온 모든 주문을 순서대로 처리해라(for)'. '재료가 떨어질 때까지 빵을 계속 구워라(while)'. **이런 식의 흐름 제어가 없으면 주방이 혼란스러워지겠죠? 딥러닝 모델도 마찬가지로 체계적인 흐름 제어가 필요해요.**

## 🚀 핵심 개념 최종 정리 (Core Concepts Final Summary)

### 01. 함수 (Function: `def`, `lambda`)
*   **핵심:** 똑같은 계산을 반복할 때, 한 번 정의해두고 필요할 때마다 재료만 바꿔서 재사용하는 자동 기계.
*   **딥러닝에서:** 딥러닝 모델은 수많은 작은 계산(함수)들이 모여서 만들어져요. 복잡한 모델을 부분별로 나누어 만들고 관리할 수 있게 해줍니다. 대량의 데이터를 한 번에 처리하는 데 사용돼요.
*   **예시:** 친구들 열 명의 시험 점수를 모두 2배로 올려주는 계산을 `점수를 2배로 만드는 함수` 하나로 처리하는 것과 같아요.

### 02. 데이터 타입 (Data Type: `int`, `float`, `list`)
*   **핵심:** 우리가 계산하고 싶은 대상(재료)의 종류나 담는 방식에 따라 다르게 다뤄야 해요. 숫자의 종류나 묶는 방식에 따라 정확한 계산을 할 수 있습니다.
*   **딥러닝에서:** 딥러닝 모델은 여러 종류의 데이터를 이해하고 처리해야 합니다. '개수'는 정수, '길이/온도'는 실수, 여러 정보는 목록으로 표현하여 모델이 데이터를 정확하게 파악하도록 돕습니다.
*   **예시:** 사람의 얼굴을 인식하려면, 얼굴 이미지를 수많은 숫자(픽셀)의 목록으로 바꿔서 모델에 넣어주는 것과 같아요.

### 03. 범위 지정 (Range: `range`)
*   **핵심:** 여러 번 반복해서 뭔가를 해야 할 때, '첫 번째부터 몇 번째까지'처럼 순서대로 정해진 횟수만큼 동작을 시키고 싶을 때 사용하는 '번호표 발급기' 역할을 합니다.
*   **딥러닝에서:** 딥러닝 모델은 데이터를 여러 번 반복해서 학습하는데, 이때 `range`를 사용해서 '몇 번 반복할지'를 정해요. 데이터의 특정 부분을 가져와 처리할 때 순서를 지정하는 데도 사용됩니다.
*   **예시:** 딥러닝 모델이 100장의 사진을 학습할 때, `range(100)`을 써서 0번째 사진부터 99번째 사진까지 순서대로 가져와 학습하라고 명령하는 것과 같아요.

### 04. 모양 변경 (Numpy Reshape: `reshape`)
*   **핵심:** 같은 숫자라도 어떻게 줄을 세우느냐에 따라 우리가 이해하는 방식이나 계산하는 방식이 달라져요. 데이터의 '배열 방식'을 바꾸는 것.
*   **딥러닝에서:** 딥러닝 모델은 입력되는 데이터의 '모양'에 매우 민감합니다. `reshape`는 어떤 형태로든 들어온 데이터를 모델이 이해할 수 있는 특정한 형태로 바꿔주는 중요한 작업입니다 (예: 이미지 데이터를 가로/세로 픽셀 형태로).
*   **예시:** 길게 늘어진 픽셀 정보를 가로 세로에 맞춰서 다시 배치하여, 딥러닝 모델이 '이건 사람 얼굴이네!' 하고 알아볼 수 있게 만드는 것과 같아요.

### 05. 정밀 계량 (arange와 linspace)
*   **핵심:** `arange`는 일정한 '간격(step)'으로 숫자를 만들고 싶을 때, `linspace`는 특정 범위 안에서 '원하는 개수(num)'만큼 숫자를 똑같이 나눠서 만들고 싶을 때 사용해요. 정밀한 숫자 생성에 필요합니다.
*   **딥러닝에서:** 딥러닝 모델에서 함수의 변화를 시각화하거나, 특정 범위 내에서 테스트할 입력 값을 생성할 때 유용해요. 활성화 함수가 입력 값에 대해 어떻게 반응하는지 연속적으로 확인하는 데 사용됩니다.
*   **예시:** 시험 점수를 0점에서 100점까지 10점 단위로 만들려면 `arange`를, 0점에서 100점까지 딱 5개의 대표값을 만들려면 `linspace`를 사용하는 것과 같아요.

### 06. 랜덤 및 시각화 (random & plot)
*   **핵심:** 우리가 모르는 새로운 상황을 예측하거나, 데이터 속에 숨어있는 규칙을 찾을 때 '무작위(random)'로 숫자를 뽑아보거나, 그 숫자들을 '그림으로 그려서(plot)' 살펴보는 것이 큰 도움이 돼요. 눈으로 봐야 패턴이나 문제점을 알 수 있습니다.
*   **딥러닝에서:** 딥러닝 모델이 처음 학습을 시작할 때, 가중치(모델의 핵심 '조절값')는 무작위로 설정되어 모델이 다양한 가능성을 탐색하게 돕습니다. 또한, 모델이 학습한 결과를 눈으로 확인하려면 그래프(플롯)가 필수적입니다.
*   **예시:** 동전 던지기 결과를 `random`으로 만들어 막대그래프나 선 그래프로 그리면, '앞면이 조금 더 많이 나왔네' 하고 한눈에 볼 수 있는 것과 같아요.

### 07. 제어문 (Control Flow: `if`, `for`, `while`)
*   **핵심:** '만약 ~라면 이렇게 하고, 아니면 저렇게 해' (`if`), '이걸 몇 번 반복해' (`for`), '이 조건이 될 때까지 계속 반복해' (`while`)처럼, 코드의 흐름을 조절하는 문법.
*   **딥러닝에서:** 딥러닝 모델의 학습 과정은 제어문의 연속입니다. 예를 들어, '학습 오차가 특정 값보다 작아지면 학습을 멈춰라' (`if`), '모든 데이터를 100번 반복해서 학습해라' (`for`), '오차가 더 이상 줄어들지 않을 때까지 학습을 계속해라' (`while`)와 같이 사용됩니다.
*   **예시:** 로봇 청소기가 '방에 먼지가 있으면 청소하고(if), 모든 방을 다 돌 때까지(for) 계속 청소해라'와 같이 동작하는 것과 같아요.
