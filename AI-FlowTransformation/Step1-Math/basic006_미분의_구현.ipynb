### 001: Symbolic Differentiation (기호적 미분)  
**수학적 의미:**  
$$ e^x \rightarrow e^x $$
$$ x^n \rightarrow nx^{n-1} $$
$$ \ln x \rightarrow \frac{1}{x} $$

**생활 비유:**  
“공식집에 있는 규칙을 그대로 적용하는 것 → 정해진 법칙을 따라 빠르게 답을 얻는 방식.”  

**딥러닝 연결:**  
신경망 학습에서 역전파(backpropagation)를 할 때, 미분 공식을 그대로 적용해 기울기를 계산합니다. 즉, **정확한 수학적 규칙 기반의 미분**이 모델 학습의 핵심입니다.

#### 🦍 **딥러닝에서의 역할 (1):**
딥러닝 모델이 학습할 때, 각 단계에서 가중치(재료)를 얼마나 조절해야 할지 정확하게 계산하는 핵심 규칙입니다. 오차를 줄이기 위한 '정확한 변화량'을 찾아냅니다.

#### 🦍 **더 쉽게 풀어보는 딥러닝 예시 (2):**
모델이 '이 상황에서는 이 재료를 이만큼만 바꿔야 최고의 맛이 나!' 하고 딱 정해진 공식을 아는 것과 같아요. 그래서 헤매지 않고 바로 최적의 조절 값을 찾아냅니다.

#### 🦍 **더더 쉽게 풀어보는 딥러닝 예시 (3):**
요리사가 '설탕 1g 늘리면 단맛이 5% 증가한다'는 정확한 과학 공식을 아는 것과 같습니다. 모델도 이런 정확한 공식을 이용해서 오차를 줄이는 최적의 길을 찾아갑니다.

### 🚀 딥러닝, 우리 주변에서 언제 사용될까요? (실생활 예시) - 기호적 미분

기호적 미분은 딥러닝 모델이 학습할 때, 각 가중치(파라미터)를 **얼마나 정확하게 조절해야 할지** 알려주는 핵심 원리입니다. 마치 요리의 정확한 레시피처럼, 모델은 이 규칙을 통해 오차를 줄이는 최적의 방향을 찾아갑니다.

*   **정확한 이미지 분류:** 자율주행차가 도로의 '정지 신호'를 정확히 인식하려면, 모델이 이미지의 미세한 패턴 변화를 **오차 없이** 감지하고 반응해야 합니다. 기호적 미분은 이러한 미세한 변화에 대한 모델의 민감도를 정확하게 계산하여, 신호등 색상이나 모양 변화에 대한 정확한 판단을 돕습니다.
*   **정교한 자연어 처리:** 챗봇이 사용자의 복잡한 질문 의도를 파악하고 적절하게 답변하려면, 단어 하나하나의 변화가 전체 문맥에 미치는 영향을 **정확한 공식**처럼 계산해야 합니다. 기호적 미분은 이러한 언어 모델의 내부 가중치가 사용자 의도에 얼마나 영향을 미치는지 정밀하게 계산하여 더 자연스러운 대화를 가능하게 합니다.

````python
import sympy
x = sympy.Symbol('x')

# 예시 1 (쉬운 예시): 숫자 3의 미분 (변화가 없으니 0)
# '변화가 없는 것'은 미분하면 '0'이 됩니다.
expr_simple = sympy.Integer(3)
diff_expr_simple = sympy.diff(expr_simple, x)
print(f"d/dx({expr_simple}) = {diff_expr_simple} (변화가 없으니 0)")

# 예시 2 (대표적인 예시): x의 미분 (x가 1만큼 변하면 결과도 1만큼 변함)
# '재료 x'가 'x'라면, 변화율은 '1'입니다. (x^1)
expr_typical = x
diff_expr_typical = sympy.diff(expr_typical, x)
print(f"d/dx({expr_typical}) = {diff_expr_typical} (x의 변화율은 1)")
````

### 002: Numerical Differentiation (수치해석적 미분)  
**수학적 의미:**  
$$ \frac{\partial f}{\partial x} \approx \frac{f(x+h) - f(x)}{h}, \quad \frac{f(x+h) - f(x-h)}{2h} $$

**생활 비유:**  
“실제 값을 직접 찍어보고 차이를 계산하는 것 → 눈금자를 대고 길이를 재는 느낌.”  

**딥러닝 연결:**  
심볼릭 미분이 어려운 경우, **근사값으로 기울기를 추정** 합니다. 특히 수치 안정성을 확인하거나, 자동미분 결과를 검증할 때 자주 활용됩니다.

#### 🦍 **딥러닝에서의 역할 (1):**
딥러닝 모델에서 정확한 미분 공식을 알기 어렵거나, 자동미분 결과가 맞는지 확인할 때 **'대략적인 기울기'** 를 계산하여 검증하는 보조적인 역할을 합니다.

#### 🦍 **더 쉽게 풀어보는 딥러닝 예시 (2):**
모델이 '음, 이 재료를 조금 더 넣었더니 맛이 이 정도 변하네? 그럼 이쪽으로 가면 더 좋겠군!' 하고 직접 살짝 변화를 줘서 맛의 변화를 대략적으로 측정하는 것과 비슷합니다.

#### 🦍 **더더 쉽게 풀어보는 딥러닝 예시 (3):**
요리사가 '레시피가 없으니, 설탕을 아주 조금 넣어보고 맛을 본 다음, 다시 소금을 조금 넣어보고 맛을 보면서 적절한 맛을 찾아가는' 과정과 같아요. 공식 대신 '경험적'으로 변화를 가늠하는 것이죠.

### 🚀 딥러닝, 우리 주변에서 언제 사용될까요? (실생활 예시) - 수치해석적 미분

수치해석적 미분은 딥러닝 모델의 복잡한 연산이 **정확하게 동작하는지 확인**하거나, 새로운 모델을 만들었을 때 **예상대로 학습되는지 검증**하는 데 사용됩니다. 직접 실험해보고 결과를 확인하는 방식과 비슷합니다.

*   **모델 디버깅 및 검증:** 딥러닝 모델이 학습 중 이상한 결과를 내거나 성능이 기대에 못 미칠 때, 개발자는 수치해석적 미분을 사용하여 모델 내부의 특정 부분(예: 특정 레이어의 가중치)이 **예상대로 기울기를 계산하고 있는지** 직접 '찍어보며' 확인합니다. 이는 마치 복잡한 기계의 부품이 제대로 작동하는지 일일이 손으로 돌려보며 점검하는 것과 같습니다.
*   **새로운 활성화 함수의 성능 테스트:** 새로운 활성화 함수(뉴런의 출력을 결정하는 함수)를 개발했을 때, 이 함수의 기울기가 이론적으로 맞는 값으로 잘 계산되는지 **초기 단계에서 수치적으로 검증**해볼 수 있습니다. 이를 통해 자동미분 시스템이 올바르게 구현되었는지 신뢰도를 높일 수 있습니다.

````python
import numpy as np

def numerical_derivative_formula(f, x_val, h=1e-4):
    # 아주 작은 변화 h를 주어 함수 값의 차이를 계산합니다.
    # (f(x+h) - f(x-h)) / (2*h)는 '중앙 차분'이라고 부르며 더 정확한 근사값을 줍니다.
    return (f(x_val + h) - f(x_val - h)) / (2 * h)

# 예시 1 (쉬운 예시): f(x) = x * 2 의 x=5에서의 기울기 추정
# '재료 x'가 1 변할 때 '결과 f(x)'는 2 변합니다. (기울기는 2)
def func_simple(x):
    return x * 2
x_val_simple = 5.0
derivative_simple = numerical_derivative_formula(func_simple, x_val_simple)
print(f"f(x) = x*2의 x={x_val_simple}에서의 기울기(추정): {derivative_simple:.4f}")

# 예시 2 (대표적인 예시): f(x) = x + 10 의 x=7에서의 기울기 추정
# '재료 x'가 1 변할 때 '결과 f(x)'는 1 변합니다. (기울기는 1)
def func_typical(x):
    return x + 10
x_val_typical = 7.0
derivative_typical = numerical_derivative_formula(func_typical, x_val_typical)
print(f"f(x) = x+10의 x={x_val_typical}에서의 기울기(추정): {derivative_typical:.4f}")
````

### 003: Numerical Differentiation with Python (파이썬 구현)  
**수학적 의미:**  
```python
def numerical_diff(f, x):
    h = 1e-4
    return (f(x+h) - f(x-h)) / (2*h)
```

**생활 비유:**  
“좌우로 아주 조금 움직여서 기울기를 측정하는 것 → 작은 움직임으로 경사 확인.”  

**딥러닝 연결:**  
실제 코드로 미분을 구현해 **함수의 기울기를 직접 계산**합니다. 이는 딥러닝 프레임워크 내부에서 자동미분이 동작하는 원리를 이해하는 데 도움을 줍니다.

#### 🦍 **딥러닝에서의 역할 (1):**
파이썬 코드로 직접 함수의 기울기(변화율)를 계산하는 방법을 배웁니다. 이는 딥러닝 프레임워크가 복잡한 미분을 어떻게 처리하는지 이해하는 기초가 됩니다.

#### 🦍 **더 쉽게 풀어보는 딥러닝 예시 (2):**
모델이 'A 재료 1개 넣었을 때 맛이 10점, 2개 넣었을 때 12점? 아하, 1개 늘릴 때마다 맛이 2점씩 좋아지는구나!' 하고 직접 해보면서 변화율을 코드로 계산하는 것입니다.

#### 🦍 **더더 쉽게 풀어보는 딥러닝 예시 (3):**
요리사가 요리책 없이 '레시피가 없으니, 내가 직접 설탕을 1g씩 추가해보면서 맛의 변화를 기록해봐야지!' 하는 실험 일지를 작성하는 것과 같아요.

### 🚀 딥러닝, 우리 주변에서 언제 사용될까요? (실생활 예시) - 파이썬으로 수치 미분 구현

파이썬으로 수치 미분을 직접 구현해보는 것은 딥러닝 프레임워크가 **복잡한 기울기를 어떻게 계산하는지** 그 '내부 작동 원리'를 이해하는 데 큰 도움을 줍니다. 마치 자동차의 엔진을 직접 조립해보면서 각 부품의 역할을 이해하는 것과 같습니다.

*   **교육용 시뮬레이션 및 개념 증명:** 딥러닝 입문자들이나 학생들이 미분 개념을 코드로 직접 체험하고 이해하기 위해 파이썬으로 간단한 수치 미분 함수를 구현해봅니다. 이를 통해 복잡한 딥러닝 라이브러리의 추상적인 개념들을 **손으로 직접 구현해보는 경험**을 할 수 있습니다.
*   **간단한 함수의 기울기 분석:** 때로는 아주 간단한 수학적 함수에 대한 기울기 변화를 시각적으로 분석할 필요가 있을 때, 복잡한 딥러닝 프레임워크 대신 파이썬으로 직접 구현한 수치 미분 함수를 사용하여 빠르게 결과를 확인하고 직관을 얻을 수 있습니다. 예를 들어, 특정 데이터 분포에 따른 손실 함수의 변화율을 이해하는 데 활용될 수 있습니다.

````python
# 003: Numerical Differentiation with Python (파이썬 구현) - 핵심 함수 정의
# 이 함수가 바로 '아주 작은 차이로 기울기를 재는 자'입니다.
def numerical_diff(f, x):
    h = 1e-4 # 아주 작은 값 (0.0001)
    # (f(x+h) - f(x-h)) / (2*h)는 '중앙 차분'으로 기울기를 근사합니다.
    return (f(x + h) - f(x - h)) / (2 * h)
````

````python
# 003: Numerical Differentiation with Python (파이썬 구현) - 예시

# 예시 1 (쉬운 예시): f(x) = x * 3 의 x=4에서의 기울기 계산
# 'x*3' 함수는 x가 1 변할 때마다 결과가 3 변합니다. (기울기는 3)
def simple_func_1(x):
    return x * 3
x_val_1 = 4.0
derivative_1 = numerical_diff(simple_func_1, x_val_1)
print(f"f(x) = x*3의 x={x_val_1}에서의 기울기: {derivative_1:.4f}")

# 예시 2 (대표적인 예시): f(x) = 5 (상수 함수) 의 x=10에서의 기울기 계산
# '결과가 항상 5로 고정'되어 있으니 x가 아무리 변해도 결과는 5입니다. 변화율은 0.
def simple_func_2(x):
    return 5 # 어떤 x가 들어와도 항상 5를 반환
x_val_2 = 10.0
derivative_2 = numerical_diff(simple_func_2, x_val_2)
print(f"f(x) = 5의 x={x_val_2}에서의 기울기: {derivative_2:.4f}")
````

### 004: Partial Derivatives (편미분)  
**수학적 의미:**  
- 일반미분: $$ \frac{df}{dx}(x_0) $$
- 편미분: $$ \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} $$

**생활 비유:**  
“여러 요인 중 하나만 바꿔서 결과가 어떻게 변하는지 보는 것 → 요리에서 소금만 조금 더 넣어 맛을 확인하는 느낌.”  

**딥러닝 연결:**  
다변수 함수(신경망)는 입력이 많습니다. **각 입력에 대한 기울기(편미분)** 를 계산해 가중치를 업데이트하는 것이 학습의 본질입니다.

#### 🦍 **딥러닝에서의 역할 (1):**
딥러닝 모델은 수많은 입력(데이터)과 가중치(조절할 값)를 가지고 있습니다. 편미분은 이 많은 것들 중에서 **딱 한 가지 요소만 움직였을 때, 결과가 얼마나 변하는지**를 알려줍니다.

#### 🦍 **더 쉽게 풀어보는 딥러닝 예시 (2):**
모델이 '이 과자의 맛은 설탕, 밀가루, 소금의 양에 따라 달라지지? 그럼 설탕만 1g 늘리면 맛이 얼마나 달라질까?' 하고 나머지 재료는 그대로 둔 채, 딱 설탕 하나만 변화시켜 보는 것과 같아요.

#### 🦍 **더더 쉽게 풀어보는 딥러닝 예시 (3):**
영화 평점에서 '배우의 연기'와 '스토리' 두 가지 요소가 중요하다고 할 때, 배우의 연기 점수만 1점 올렸을 때 영화 전체 평점이 얼마나 오를지 측정하는 것과 같습니다. 다른 요소는 고정하고 하나만 집중해서 보는 것이죠.

### 🚀 딥러닝, 우리 주변에서 언제 사용될까요? (실생활 예시) - 편미분

딥러닝 모델은 수많은 '가중치(Weight)'와 '편향(Bias)'이라는 조절 가능한 변수들을 가지고 있습니다. 편미분은 이 수많은 변수들 중에서 **'딱 하나의 변수만 조절했을 때 모델의 성능(오차)이 얼마나 변할지'** 알려주는 핵심 도구입니다.

*   **신경망 가중치 최적화:** 이미지 분류 모델이 '고양이' 사진을 '개'라고 잘못 분류했을 때, 모델은 수십만, 수백만 개의 가중치 중에서 **어떤 가중치를 얼마나 변경해야** 이 오류를 줄일 수 있을지 편미분을 통해 계산합니다. 각 가중치가 전체 오차에 미치는 영향(기울기)을 편미분으로 파악하여 업데이트 방향을 결정합니다.
*   **추천 시스템 개인화:** 넷플릭스 추천 시스템에서 '특정 영화 장르에 대한 선호도'라는 변수 하나가 사용자에게 추천될 다음 영화에 얼마나 큰 영향을 미치는지 편미분을 통해 분석할 수 있습니다. 다른 모든 변수는 고정한 채, 이 장르 선호도 변수만 변경했을 때 추천 결과가 어떻게 달라지는지 파악하여 **더 정교한 개인화**를 제공합니다.

````python
import sympy
x, y = sympy.symbols('x y') # 변수 x, y를 사용하겠다고 '심볼'로 정의합니다.

# 예시 1 (쉬운 예시): f(x, y) = x + y 에서 x에 대한 변화만 보기
# '요리'는 '사과 개수(x) + 바나나 개수(y)'. 사과 개수만 1개 늘리면 전체 요리는 1만큼 변합니다.
func_add = x + y
partial_diff_x_add = sympy.diff(func_add, x) # x에 대해서만 미분
print(f"f(x, y) = {func_add}를 x에 대해 편미분: {partial_diff_x_add}")

# 예시 2 (대표적인 예시): f(x, y) = x * y 에서 x에 대한 변화만 보기
# '요리'는 '설탕 양(x) * 소금 양(y)'. 소금 양(y)이 고정된 상태에서 설탕 양(x)을 늘리면,
# 요리의 변화는 소금 양(y)만큼 발생합니다.
func_mul = x * y
partial_diff_x_mul = sympy.diff(func_mul, x) # x에 대해서만 미분
print(f"f(x, y) = {func_mul}를 x에 대해 편미분: {partial_diff_x_mul}")
````

### 005: Tangent Line (접선 구하기)

직선의 기울기: $$ \frac{y_2 - y_1}{x_2 - x_1} $$
접선 방정식: $$ y = ax + b $$

**생활 비유:**  
“길 위에서 순간의 방향을 알려주는 나침반 → 현재 위치에서 앞으로 나아갈 방향.”  

**딥러닝 연결:**  
미분은 결국 **접선의 기울기** 를 구하는 과정입니다. 이 기울기가 경사하강법에서 파라미터 업데이트의 방향을 결정합니다.

#### 🦍 **딥러닝에서의 역할 (1):**

#### 🦍 **더 쉽게 풀어보는 딥러닝 예시 (2):**

### 006: Tangent Line in Python (접선 방정식 구현)  
**수학적 의미:**  
```python
def tangent_line(f, x1):
    y1 = f(x1)
    a = numerical_diff(f, x1)
    b = y1 - a*x1
    return lambda x: a*x + b
```

**생활 비유:**  
“현재 위치에서 직선 길을 그려 앞으로 나아갈 길을 표시하는 것.”  

**딥러닝 연결:**  
접선은 **기울기 기반 최적화**의 직관적 표현입니다. 모델은 접선의 방향을 따라 파라미터를 조정하며 손실을 줄여갑니다.

#### 🦍 **딥러닝에서의 역할 (1):**
파이썬 코드를 이용해 특정 지점에서의 '최적의 하강 방향'을 나타내는 접선을 실제로 그려봅니다. 이는 모델이 오차를 줄이기 위해 가중치를 어떻게 조절하는지 시각적으로 이해하는 데 도움을 줍니다.

#### 🦍 **더 쉽게 풀어보는 딥러닝 예시 (2):**
모델이 '지금 맛이 너무 짜니까, 소금을 줄이는 방향으로 조절해야겠어. 그럼 소금 양을 얼마나 줄여야 가장 적절할까?' 하고 그 조절의 '방향과 강도'를 코드로 계산하고 표시하는 것과 같아요.

#### 🦍 **더더 쉽게 풀어보는 딥러닝 예시 (3):**
주차를 할 때 핸들을 어느 방향으로 얼마나 돌려야 원하는 곳에 정확히 주차할 수 있는지, 그 움직임의 '가이드 라인'을 코드로 만들어 보는 것과 같습니다.

````python
# 006: Tangent Line in Python (접선 방정식 구현) - 핵심 함수 정의
# 이 함수는 특정 지점에서의 '순간적인 변화 방향 길' (접선)을 찾아주는 '길잡이'입니다.
# 'numerical_diff' 함수(003번 셀)가 필요합니다.
def tangent_line(f, x1):
    y1 = f(x1) # 현재 지점의 y 값
    a = numerical_diff(f, x1) # 현재 지점에서의 순간 기울기 (접선의 기울기)
    b = y1 - a * x1 # 접선 방정식 y = ax + b 에서 b 값 계산
    return lambda x: a * x + b # 접선 방정식 함수를 반환
````

````python
# 006: Tangent Line in Python (접선 방정식 구현) - 예시

# 예시 1 (쉬운 예시): f(x) = 2*x + 3 의 x=1에서의 접선 방정식 구하기
# 이 함수는 이미 직선이므로, 어떤 점에서든 접선은 자기 자신과 같습니다.
def func_linear(x):
    return 2 * x + 3
x_point_1 = 1.0
tangent_func_1 = tangent_line(func_linear, x_point_1)

print(f"f(x)=2x+3의 x={x_point_1}에서의 접선: y = {numerical_diff(func_linear, x_point_1):.2f}x + {tangent_func_1(0):.2f}")
print(f" (이 경우, 함수 자체가 직선이라 접선이 함수 자신입니다!)")

# 예시 2 (대표적인 예시): f(x) = x**2 의 x=2에서의 접선 방정식 구하기
# 곡선 위에 있는 특정 한 점에서의 '순간적인 방향'을 나타내는 직선을 찾아봅니다.
def func_quadratic(x):
    return x**2
x_point_2 = 2.0
tangent_func_2 = tangent_line(func_quadratic, x_point_2)

# x=2.5 일 때 접선 위의 y 값 계산 (실제 곡선과는 다름)
y_on_tangent_2 = tangent_func_2(2.5)
print(f"\nf(x)=x^2의 x={x_point_2}에서의 접선: y = {numerical_diff(func_quadratic, x_point_2):.2f}x + {tangent_func_2(0):.2f}")
print(f"접선 위의 x=2.5 일 때의 y값: {y_on_tangent_2:.2f}")
````

### 007: Computational Graph (계산 그래프)  
**수학적 의미:**  
복잡한 함수 $$ f(x,y) $$를 연산 노드(+, ×, 상수, 변수)로 나눠 표현.  

**생활 비유:**  
“큰 요리를 레시피 단계별로 나누는 것 → 재료와 과정이 연결된 흐름.”  

**딥러닝 연결:**  
딥러닝 프레임워크는 계산 그래프를 통해 **자동미분**을 수행합니다. 각 노드에서 미분을 계산하고 연결해 전체 기울기를 구합니다.

#### 🦍 **딥러닝에서의 역할 (1):**
딥러닝 모델의 복잡한 계산 과정을 마치 '레시피 카드'처럼 단계별로 쪼개어 그림으로 보여주는 것입니다. 각 단계(노드)는 작은 계산(예: 더하기, 곱하기)을 담당하고, 이들이 연결되어 최종 결과가 나옵니다.

#### 🦍 **더 쉽게 풀어보는 딥러닝 예시 (2):**
모델이 '이 음식은 밥을 짓고 → 재료를 썰고 → 양념을 섞고 → 볶는' 여러 단계를 거쳐 완성돼! 하고 각 단계를 명확히 분리해서 보여주는 것과 같습니다.

#### 🦍 **더더 쉽게 풀어보는 딥러닝 예시 (3):**
복잡한 조립식 가구를 만들 때, '1단계: 다리 조립 → 2단계: 상판 연결 → 3단계: 나사 조이기' 처럼 각 과정을 순서대로 그림으로 표시한 설명서와 같아요. 모델도 이 그림을 따라 복잡한 계산을 합니다.

### 🚀 딥러닝, 우리 주변에서 언제 사용될까요? (실생활 예시) - 파이썬으로 접선 방정식 구현

파이썬으로 접선 방정식을 구현해보는 것은 딥러닝 모델이 학습 과정에서 **'어떤 길을 따라 최적화되어 가는지'**를 시각적으로 이해하는 데 유용합니다. 실제 모델은 이 접선 방향을 따라 가중치를 조금씩 조절하며 성능을 개선합니다.

*   **모델 학습 시각화:** 딥러닝 교육 자료나 연구에서 손실 함수(Loss Function)의 변화를 설명할 때, 특정 지점에서의 '접선'을 파이썬으로 그려서 **경사 하강법(Gradient Descent)이 어떻게 동작하는지**를 직관적으로 보여줍니다. 예를 들어, 2차원 평면에서 손실 함수의 최저점을 찾아가는 과정에서 각 스텝마다 모델이 어떤 방향으로 움직이는지 접선으로 표현하여 이해를 돕습니다.
*   **최적화 알고리즘 테스트:** 새로운 최적화 알고리즘(예: Adam, RMSprop 등)을 개발하거나 기존 알고리즘의 동작 방식을 이해할 때, 간단한 함수에 대해 파이썬으로 접선을 계산하고 이를 시각화하여 **알고리즘이 올바른 '하강 방향'을 찾고 있는지** 테스트하고 검증하는 데 활용할 수 있습니다.

````python
# 예시 1 (쉬운 예시): '2 + 3' 계산을 그래프로 표현
# 마치 레시피 '사과 2개 + 바나나 3개' 처럼 단계별로 봅니다.
a = 2         # '사과 2개' 입력 노드
b = 3         # '바나나 3개' 입력 노드
result = a + b # '사과+바나나' 덧셈 노드

print(f"입력: 사과={a}, 바나나={b}")
print(f"계산 단계: 사과 + 바나나 = {result}")

# 예시 2 (대표적인 예시): '(5 * 2) + 1' 계산을 그래프로 표현
# '스테이크 5개 * 가격 2'를 먼저 계산하고 '추가 소스 1'을 더합니다.
steak_count = 5
price_per_steak = 2
additional_sauce = 1

# 1단계: 스테이크 총 가격 계산 (곱셈 노드)
total_steak_price = steak_count * price_per_steak
# 2단계: 최종 요금 계산 (덧셈 노드)
final_bill = total_steak_price + additional_sauce

print(f"\n스테이크 {steak_count}개, 개당 {price_per_steak}원, 소스 {additional_sauce}원")
print(f"1단계 (곱셈): 스테이크 총 가격 = {total_steak_price}")
print(f"2단계 (덧셈): 최종 요금 = {final_bill}")
````

### 🚀 딥러닝, 우리 주변에서 언제 사용될까요? (실생활 예시) - 계산 그래프

계산 그래프는 딥러닝 모델의 복잡한 연산 과정을 **'단계별 레시피'처럼 시각적으로 표현**한 것입니다. 이 구조 덕분에 모델은 수백, 수천 개의 연산을 효율적으로 관리하고, 각 단계에서 필요한 미분값(기울기)을 자동으로 계산할 수 있습니다.

*   **복잡한 딥러닝 모델 설계:** 이미지 인식 모델(예: CNN)이나 자연어 처리 모델(예: Transformer)과 같이 수십, 수백 개의 층(Layer)과 다양한 연산으로 구성된 모델을 설계할 때, 계산 그래프는 각 층의 연산이 어떻게 연결되고 데이터가 어떻게 흘러가는지 **명확한 설계도 역할**을 합니다. 이를 통해 모델의 오류를 찾고 성능을 최적화하는 데 도움이 됩니다.
*   **자동미분을 통한 효율적인 학습:** 딥러닝 프레임워크(TensorFlow, PyTorch 등)는 이 계산 그래프를 기반으로 **자동미분(Automatic Differentiation)**을 수행합니다. 즉, 모델 개발자가 일일이 미분 공식을 작성할 필요 없이, 그래프를 따라 역방향으로 각 노드의 미분값을 효율적으로 계산하여 학습 속도를 비약적으로 높입니다. 이는 마치 복잡한 세금 계산을 사람이 일일이 하지 않고, 전문 소프트웨어가 자동으로 처리해주는 것과 같습니다.

````python
import tensorflow as tf

# 예시 1 (쉬운 예시): TensorFlow로 간단한 덧셈 계산하기
# '빵 1개 + 우유 2개'를 계산한다고 상상해 보세요.
a_item = tf.constant(1) # 빵 1개
b_item = tf.constant(2) # 우유 2개

# 두 숫자를 더하는 연산
total_items = a_item + b_item

print(f"빵 개수 (a): {a_item.numpy()}")
print(f"우유 개수 (b): {b_item.numpy()}")
print(f"총 개수 (a + b): {total_items.numpy()}")

# 예시 2 (대표적인 예시): TensorFlow로 '재료 양 변화에 따른 맛 변화' 계산하기 (자동 미분)
# '맛' 함수가 '재료 양의 제곱'이라고 할 때, 재료 양을 '어떻게 바꿔야' 맛이 가장 빨리 변할까요?
x_ingredient = tf.Variable(3.0) # 현재 재료 양 (바꿀 수 있는 값)

with tf.GradientTape() as tape:
    # '맛' = '재료 양' * '재료 양' (x*x)
    y_taste = x_ingredient ** 2

# '맛' (y_taste)이 '재료 양' (x_ingredient)에 대해 얼마나 변하는지 (기울기) 계산
gradient_taste = tape.gradient(y_taste, x_ingredient)

print(f"\n현재 재료 양 (x): {x_ingredient.numpy()}")
print(f"현재 맛 (y = x^2): {y_taste.numpy()}")
print(f"재료 양에 대한 맛의 변화율 (기울기, 2x): {gradient_taste.numpy()}")
````

## 🚀 핵심 개념 최종 정리

### 01. 기호적 미분 (Symbolic Differentiation)
*   **핵심:** 정해진 **수학 공식(레시피)을 그대로 사용**해서 **정확한 변화율**을 찾는 방식.
*   **딥러닝에서:** 모델의 가중치를 업데이트할 때 **가장 기본이 되는 정확한 기울기 계산** 방법이에요.

### 02. 수치해석적 미분 (Numerical Differentiation)
*   **핵심:** 실제 값을 **조금씩 바꿔보며(직접 측정) 대략적인 변화율**을 추정하는 방식.
*   **딥러닝에서:** 기호적 미분 결과가 맞는지 확인하거나 검증하는 보조 수단으로 사용돼요.

### 03. 편미분 (Partial Derivatives)
*   **핵심:** 여러 요인 중 **'딱 하나의 요인만 바꿔봤을 때' 변화가 얼마나 생기는지** 알아보는 방식.
*   **딥러닝에서:** 수많은 가중치 중에서 **'각각의 가중치가 모델 성능에 얼마나 영향을 주는지'** 를 파악해서 업데이트 방향을 정해요.

### 04. 접선 구하기 (Tangent Line)
*   **핵심:** 특정 지점에서 **'순간적으로 어느 방향으로 나아가야 할지'** 를 알려주는 길(방향)을 찾는 것.
*   **딥러닝에서:** 현재 모델의 상태에서 **오차를 가장 빠르게 줄일 수 있는 '가중치 업데이트 방향'** 을 결정하는 데 쓰여요.
