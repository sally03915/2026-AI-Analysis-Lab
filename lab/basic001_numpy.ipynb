#### 1. 셰프의 마법 도구: 함수 만들기

딥러닝은 수많은 함수의 연결입니다. 재료를 넣으면 결과를 뱉어내는 자동 기계를 만들어 봅시다.

```python
# [def] 정식 셰프: 복잡한 로직을 담을 때
def power_up(x):
    return x**2 + 10

# [lambda] 패스트푸드: 한 줄로 끝내는 간편함
fast_power = lambda x: x**2 + 10

print(f"def 사용: {power_up(5)}, lambda 사용: {fast_power(5)}")
```

#### 2. 재료 손질하기 (데이터 생성)

컴퓨터가 알아듣게 정수(`int`), 실수(`float`), 목록(`list`)으로 재료를 다듬습니다.

```python
# 정수(int): 77을 세 가지 방식으로 생성
A1 = 77; A2 = int('77'); A3 = int()
print('A1:', A1, 'A2:', A2, 'A3:', A3)

# 실수(float): 3.4를 세 가지 방식으로 생성
B1 = 3.4; B2 = float('3.4'); B3 = float()
print('B1:', B1, 'B2:', B2, 'B3:', B3)

# 목록(list): 여러 재료를 바구니에
C1 = [0,1,2,3];   C2 = list(range(4)); C3 = list()
print('C1:', C1, 'C2:', C2, 'C3:', C3)
```

#### 3. 재료 번호표 뽑기 (range)

재료가 많을 땐 `range`로 번호표를! **주의: 마지막 숫자는 절대 포함 안 해요!**

```python
# range(숫자1) - 시작 0, 끝 5 직전까지
print(list(range(5)))        # [0, 1, 2, 3, 4]
# range(숫자1, 숫자2) - 시작 1, 끝 5 직전까지
print(list(range(1, 5)))     # [1, 2, 3, 4]
# range(숫자1, 숫자2, 숫자3) - 시작 1, 끝 5 직전까지, 스텝 2
print(list(range(1, 5, 2)))  # [1, 3]
print(list(range(1, 6, 2)))  # [1, 3, 5]
```

#### 4. 모양 맞추기: Numpy Reshape

똑같은 24개의 재료라도 어떤 접시(Shape)에 담느냐에 따라 요리 모양이 달라집니다.

```python
A = np.array(range(24))
print("기본값" , A)
print("shape" , A.shape)
print("\n\n")

B = A.reshape(2, 12) # 2단 12열
C = A.reshape(3, 8)  # 3단 8열
D = A.reshape(4, -1) # 4단, 나머지는 알아서 채워줘!
print("24개 재료의 변신:\n", B, "\n\n", C, "\n\n", D)
```

#### 5. 정밀 계량: arange와 linspace

`arange`는 **간격(step)** 중시, `linspace`는 **개수(num)** 중시!

```python
# arange: 1부터 10 전까지 2씩 띄어서
A1 = np.arange(1, 10, 2)
# linspace: 1부터 9까지를 똑같은 5등분으로
A2 = np.linspace(1, 9, 5)

A3 = np.linspace(1,10,5)
print('간격 중심(arange):', A1, '\n\n개수 중심(linspace):', A2  ,'\n\n개수 중심(linspace):', A3)
```

##### 1. '구간'과 '숫자'의 차이

`np.linspace(1, 10, 5)`는 "1부터 10까지 5개의 숫자를 찍어라"는 뜻입니다.

* **숫자 5개를 찍으려면:** 1(첫 번째), ?, ?, ?, 10(다섯 번째)
* **사이사이의 틈(구간)은 몇 개가 생기나요?**
* 1번과 2번 사이(1구간)
* 2번과 3번 사이(2구간)
* 3번과 4번 사이(3구간)
* 4번과 5번 사이(4구간)
* **결과적으로 4개의 구간**이 생깁니다.



---

##### 2. 왜 (5 - 1)로 나누나요?

전체 거리(10 - 1 = 9)를 똑같은 크기의 '막대기' 4개로 쪼개야 하니까, **전체 길이 9를 4로 나누는 것**입니다.

* **수식:** $\text{간격} = \frac{\text{끝값} - \text{시작값}}{\text{개수} - 1}$
* **대입:** $\frac{10 - 1}{5 - 1} = \frac{9}{4} = 2.25$

그래서 숫자가 1에서 시작해서 2.25씩 커지는 것입니다.

* 1번째 숫자: **1.0**
* 2번째 숫자: 1.0 + 2.25 = **3.25**
* 3번째 숫자: 3.25 + 2.25 = **5.5**
* 4번째 숫자: 5.5 + 2.25 = **7.75**
* 5번째 숫자: 7.75 + 2.25 = **10.0**

#### 6. 랜덤 시즈닝과 맛 평가 (random & plot)

때로는 무작위(`random`) 시즈닝이 창의적인 요리를 만듭니다. 정규분포(`normal`)를 쓰면 데이터가 균형 잡힌 맛을 내죠.

```python
# 랜덤 데이터 생성
A1 = np.random.random(100); A2 = np.random.rand(100)
A1.sort(); A2.sort()
plt.plot(A1, c='b'); plt.plot(A2, c='r')
plt.show()

# 정규분포 데이터 (평균 -0.5, 편차 0.3, 250개)
data_norm = np.random.normal(-0.5, 0.3, 250)
data_norm.sort()

# 결과 시각화
plt.scatter(data_norm, data_norm, c='b', marker='.')
plt.show()
plt.plot(data_norm, c='b')
plt.show()
```


#### 1. 왜 중요한가요? (데이터의 '성질'을 파악하는 눈)

현실 세계의 데이터는 무작위로 흩어져 있습니다. 우리가 가진 코드는 "무작위 데이터 속에 숨겨진 규칙(패턴)을 찾아내는 연습"을 하는 것입니다.

* **무작위 데이터(`random`)**: 0과 1 사이에서 아무 숫자나 뽑아보면, 처음엔 뒤죽박죽입니다. 하지만 `sort()`로 정렬하고 그래프를 그리면, "아, 데이터가 이렇게 골고루 퍼져 있구나"라는 **분포의 특징**이 한눈에 들어옵니다.
* **정규분포(`normal`)**: 자연계의 거의 모든 데이터(키, 몸무게, 시험 점수 등)는 정규분포(종 모양 곡선)를 따릅니다. 이 분포를 이해하면, 미래에 어떤 데이터가 들어왔을 때 "이건 평범한 데이터다" 혹은 "이건 아주 특이한 데이터(이상치)다"를 판단하는 **통계적 기준**을 세울 수 있습니다.

#### 2. 왜 쉬운가요? (시각화의 마법)

데이터를 눈으로 보지 않으면 그냥 '숫자 덩어리'일 뿐입니다. 하지만 `matplotlib`(`plt`)을 쓰면 복잡한 통계 계산을 **단 3줄의 코드로 그림**으로 바꿀 수 있습니다.

* `plt.plot`: 데이터가 전체적으로 어떻게 흐르는지(추세) 보여줍니다.
* `plt.scatter`: 데이터 하나하나가 어디에 찍히는지(밀도) 보여줍니다.
* **재미 포인트:** 그냥 숫자만 출력하면 지루하지만, `plt.show()`를 누르는 순간 **내가 만든 데이터가 예술적인 그래프로 변하는 쾌감**이 있습니다. 마치 데이터에 생명을 불어넣는 느낌이죠!
