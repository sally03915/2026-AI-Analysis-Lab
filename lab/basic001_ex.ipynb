### [문제 1] 함수 만들기 (def vs lambda)

입력값 $x$에 3을 곱하고 5를 더하는 함수를 만들고 싶습니다.

1. `def`를 사용하여 `triple_add`라는 이름의 함수를 만드세요.
2. `lambda`를 사용하여 같은 기능을 하는 `fast_triple`이라는 변수를 만드세요.
3. 두 함수에 각각 `10`을 넣었을 때의 결과값을 출력하세요.

```python
def triple_add(x): return x * 3 + 5
fast_triple = lambda x: x * 3 + 5
print(triple_add(10), fast_triple(10)) # 결과: 35, 35
```

### [문제 3] 모양 맞추기 (Numpy Reshape)

`data = np.arange(12)`가 있습니다. 이 12개의 재료를 가지고 다음 모양의 배열(ndarray)을 만드세요.

1. 3행 4열 (3x4)
2. 6행 2열 (6x2)
3. 2행 6열 (2x6)

```python
data = np.arange(12)
print(data.reshape(3, 4))
print(data.reshape(6, 2))
print(data.reshape(2, 6))
```

### [문제 4] 정밀 계량 (`arange` vs `linspace`)

다음 두 경우의 차이를 이해했는지 확인하는 문제입니다.

1. `np.arange(0, 10, 3)`의 결과값은 무엇일까요? (실행하지 말고 먼저 예측해보세요!)
2. `np.linspace(0, 10, 6)`를 사용하면 총 몇 개의 숫자가 생성되며, 숫자와 숫자 사이의 간격(Step)은 얼마일까요?

1. `[0, 3, 6, 9]` (0부터 10 미만까지 3씩 증가)
2. 6개의 숫자가 생성됨. 간격: $(10 - 0) / (6 - 1) = 10 / 5 = 2$

### [문제 5] 응용: 셰프의 계산기

`np.linspace(0, 20, 5)`를 사용하면 생성되는 5개의 숫자를 직접 계산해 보세요.

* 시작값: 0, 끝값: 20, 개수: 5
* 힌트: 앞서 배운 공식 $\text{간격} = \frac{\text{끝값} - \text{시작값}}{\text{개수} - 1}$을 활용하세요.

간격 = $(20 - 0) / (5 - 1) = 20 / 4 = 5$
결과: `[0.0, 5.0, 10.0, 15.0, 20.0]`
